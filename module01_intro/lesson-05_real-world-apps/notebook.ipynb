### Live Demo — Simple Recommender

This is a simplified version of **content-based filtering** — recommending items similar to what a user liked. You'll build the full version in Module 14.

````python
# Mini content-based recommender — no libraries needed
# Idea: each product has tags. Similarity = how many tags they share.

# Product catalogue: name → set of tags
products = {
    "Wireless Earbuds"     : {"electronics", "audio", "wireless", "portable"},
    "Bluetooth Speaker"    : {"electronics", "audio", "wireless", "home"},
    "Laptop Stand"         : {"electronics", "accessories", "office", "portable"},
    "Mechanical Keyboard"  : {"electronics", "accessories", "office", "gaming"},
    "Gaming Mouse"         : {"electronics", "accessories", "gaming", "wireless"},
    "Yoga Mat"             : {"fitness", "home", "portable"},
    "Running Shoes"        : {"fitness", "outdoor", "sports"},
    "Water Bottle"         : {"fitness", "outdoor", "portable", "sports"},
    "Desk Lamp"            : {"home", "office", "accessories"},
    "Noise Cancelling Headphones": {"electronics", "audio", "wireless", "office", "portable"},
}

def jaccard_similarity(set_a, set_b):
    """Similarity = shared tags / all tags combined. Range: 0 (no match) to 1 (identical)."""
    intersection = len(set_a & set_b)
    union        = len(set_a | set_b)
    return intersection / union if union > 0 else 0

def recommend(liked_product, top_n=3):
    liked_tags = products[liked_product]
    scores = []
    for name, tags in products.items():
        if name == liked_product:
            continue
        sim = jaccard_similarity(liked_tags, tags)
        scores.append((name, sim))
    scores.sort(key=lambda x: -x[1])
    return scores[:top_n]


# Test the recommender
test_products = ["Wireless Earbuds", "Yoga Mat", "Mechanical Keyboard"]

for product in test_products:
    recs = recommend(product)
    print(f"Because you viewed: '{product}'")
    print(f"  You might also like:")
    for rec_name, score in recs:
        print(f"    • {rec_name:<35} (similarity: {score:.2f})")
    print()
````

**What the similarity score means:** A score of `0.50` means half the tags overlap. The higher the score, the more similar the products.

Real recommenders at Flipkart or Amazon use millions of products, billions of user interactions, and much more sophisticated algorithms — but the core idea of **finding similarity** is the same.

### Live Demo — Delivery Time Estimator

````python
# Mini delivery time estimator — shows how multiple features combine into one prediction

def estimate_delivery_time(order):
    """
    Estimates delivery time in minutes based on order features.
    Real Zomato/Swiggy models use ML with thousands of features;
    this shows the same multi-factor thinking in plain Python.
    """

    # Base: restaurant preparation time
    prep_time = order["prep_time_mins"]

    # Distance component: assume agent travels at 20 km/h on average
    distance_km   = order["distance_km"]
    travel_time   = (distance_km / 20) * 60  # convert hours → minutes

    # Traffic multiplier by time of day
    hour = order["hour_of_day"]
    if   7  <= hour <= 10: traffic_factor = 1.6   # morning rush
    elif 12 <= hour <= 14: traffic_factor = 1.3   # lunch rush
    elif 18 <= hour <= 21: traffic_factor = 1.8   # evening peak
    else                 : traffic_factor = 1.0   # off-peak

    # Weather adjustment
    weather_penalty = {"clear": 0, "cloudy": 3, "rain": 10, "heavy_rain": 20}
    weather_delay   = weather_penalty.get(order["weather"], 0)

    # Agent availability penalty
    agent_wait = max(0, (order["nearby_agents"] - 1) * -2 + 5)  # fewer agents → longer wait
    agent_wait = min(agent_wait, 12)                              # cap at 12 mins

    # Final estimate
    total = (prep_time + travel_time * traffic_factor
             + weather_delay + agent_wait)
    return round(total)


# Test with different scenarios
orders = [
    {"label": "Weekday lunch, nearby, clear",
     "prep_time_mins": 15, "distance_km": 2.5, "hour_of_day": 13,
     "weather": "clear",      "nearby_agents": 5},

    {"label": "Evening peak, far, rain",
     "prep_time_mins": 20, "distance_km": 6.0, "hour_of_day": 19,
     "weather": "rain",       "nearby_agents": 2},

    {"label": "Late night, close, clear, few agents",
     "prep_time_mins": 10, "distance_km": 1.0, "hour_of_day": 23,
     "weather": "clear",      "nearby_agents": 1},

    {"label": "Morning rush, medium distance, cloudy",
     "prep_time_mins": 18, "distance_km": 4.0, "hour_of_day": 8,
     "weather": "cloudy",     "nearby_agents": 3},
]

print(f"{'Scenario':<45} {'ETA':>5}")
print("-" * 55)
for order in orders:
    eta = estimate_delivery_time(order)
    print(f"{order['label']:<45} {eta:>3} min")
````

**The pattern here is everywhere in data science:**  
Take multiple input features → apply weights/rules/a model → produce one useful number.

The real Zomato model does this with gradient boosting trained on 100 million past deliveries — but the logic is structurally identical to what you just ran.

## 10. Finance — Credit Scoring — Code Demo

### The problem
Should a bank give this person a home loan? The answer determines whether a family gets a home or doesn't. The model that makes this call needs to be accurate, fair, and fast.

### Live Demo — Credit Risk Scorer

This is a simplified version of how a CIBIL-style score is computed. Real systems use hundreds of features and machine learning models trained on millions of accounts — but the core logic of combining weighted signals is the same.

````python
# Mini credit risk scorer
# Each feature contributes points toward a 300–900 score (like CIBIL)

def compute_credit_score(applicant):
    score = 300  # minimum base score
    breakdown = {}

    # 1. Payment history (35% weight — most important factor)
    payment_pts = {
        "always_on_time"   : 210,
        "1_2_late"         : 140,
        "several_late"     : 70,
        "defaulted"        : 0,
    }.get(applicant["payment_history"], 0)
    score += payment_pts
    breakdown["Payment history"] = payment_pts

    # 2. Credit utilisation — how much of the credit limit is being used
    utilisation = applicant["credit_utilisation_pct"]
    if   utilisation <= 10: util_pts = 120
    elif utilisation <= 30: util_pts = 90
    elif utilisation <= 50: util_pts = 60
    elif utilisation <= 75: util_pts = 30
    else                  : util_pts = 0
    score += util_pts
    breakdown["Credit utilisation"] = util_pts

    # 3. Credit history length
    years = applicant["credit_history_years"]
    age_pts = min(years * 6, 90)   # 6 points per year, capped at 90
    score += age_pts
    breakdown["Credit history age"] = round(age_pts)

    # 4. Types of credit (mix of home loan, car loan, credit card = good)
    credit_mix_pts = applicant["num_credit_types"] * 30
    credit_mix_pts = min(credit_mix_pts, 90)
    score += credit_mix_pts
    breakdown["Credit mix"] = credit_mix_pts

    # 5. Recent credit enquiries (too many = desperate for credit = risk signal)
    enquiries     = applicant["hard_enquiries_last_year"]
    enquiry_pts   = max(0, 90 - enquiries * 20)
    score        += enquiry_pts
    breakdown["Recent enquiries"] = enquiry_pts

    final_score = min(int(score), 900)

    # Risk category
    if   final_score >= 750: risk = "Excellent — likely approved at best rates"
    elif final_score >= 650: risk = "Good — approved, standard rates"
    elif final_score >= 550: risk = "Fair — may be approved with conditions"
    else                   : risk = "Poor — likely rejected"

    return final_score, risk, breakdown


# Test with three applicant profiles
applicants = [
    {
        "name"                    : "Priya — responsible borrower",
        "payment_history"         : "always_on_time",
        "credit_utilisation_pct"  : 22,
        "credit_history_years"    : 8,
        "num_credit_types"        : 3,
        "hard_enquiries_last_year": 0,
    },
    {
        "name"                    : "Rohan — new to credit",
        "payment_history"         : "1_2_late",
        "credit_utilisation_pct"  : 60,
        "credit_history_years"    : 1,
        "num_credit_types"        : 1,
        "hard_enquiries_last_year": 4,
    },
    {
        "name"                    : "Karan — rebuilding after default",
        "payment_history"         : "several_late",
        "credit_utilisation_pct"  : 85,
        "credit_history_years"    : 5,
        "num_credit_types"        : 2,
        "hard_enquiries_last_year": 2,
    },
]

for applicant in applicants:
    score, risk, breakdown = compute_credit_score(applicant)
    print(f"{'=' * 55}")
    print(f"Applicant : {applicant['name']}")
    print(f"Score     : {score} / 900")
    print(f"Assessment: {risk}")
    print(f"Breakdown :")
    for factor, pts in breakdown.items():
        bar = "█" * (pts // 10)
        print(f"  {factor:<25} {bar} +{pts}")
    print()
````

**Why this demo matters:** A credit score is not magic — it's a weighted combination of signals, translated into a number. Understanding this helps you:
1. See how features become predictions (a core data science concept)
2. Spot potential unfairness — someone "new to credit" may be penalised simply for having a short history, not for being irresponsible
3. Understand that every ML model output is explainable if you know the features

## 13. The Dark Side — Ethics and Risks

Every powerful technology has risks. Data science is no exception. As a future data professional, you're responsible for thinking about these — not just building what's technically possible.

### Real problems you need to know about

**Bias in models**  
A hiring algorithm trained on 10 years of historical data learns that the company mostly hired men for technical roles. Now it automatically ranks women lower — not because of code, but because of biased training data. The model reflects the past, not what should happen.

**Privacy violations**  
The retailer pregnancy prediction story from Lesson 3 is an example. Using personal data in ways people didn't consent to — even if it's technically legal — erodes trust and causes real harm.

**Opaque decisions**  
A loan rejection driven by an ML model with no explanation is not just frustrating — in many countries it's increasingly illegal. "The model said no" is not a valid explanation for denying someone credit.

**Feedback loops**  
A crime prediction model sends more police to certain neighbourhoods → more arrests happen there → more crime data from that neighbourhood → the model concludes it's even higher risk → even more police. The model amplifies existing patterns instead of finding truth.

### The questions you should always ask
1. **Who could be harmed by this model being wrong?**
2. **Is the training data representative of all groups affected?**
3. **Can the output be explained to the person it affects?**
4. **Who is accountable when the model makes a mistake?**

These questions will appear in every module of this course. Data science without ethics is just a sophisticated way to make the same old mistakes faster.

## ⚠️ Common Misconceptions

### Misconception 1: "Data science only works in tech companies"
**Wrong.** Agriculture, healthcare, government, sports, and manufacturing all use data science — and these sectors employ far more people than tech in India. The *application* of data science in non-tech domains often has a higher real-world impact.

### Misconception 2: "These applications are built by very specialised experts"
**Mostly wrong.** Most of the applications in this lesson are built with the same tools you'll learn in this course — Python, Pandas, Scikit-learn, SQL. The fraud detection demo, the delivery estimator, and the recommender all run on the skills from Modules 2–12. Specialised deep learning is only needed for image/speech/language tasks.

### Misconception 3: "If an AI made the decision, it must be objective"
**Wrong.** AI models are trained on human-generated data, designed by humans, and evaluated by humans. Every bias that humans have can be encoded into a model — and sometimes amplified. A model is only as objective as its training data and design choices.
