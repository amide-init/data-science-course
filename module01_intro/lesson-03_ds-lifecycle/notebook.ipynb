## 1. The 7-Step Lifecycle

Every data science project — from a student assignment to a production system at Google — moves through the same seven stages:

```
  ┌─────────────────────────────────────────────────────────────┐
  │                                                             │
  │   1. DEFINE        What problem are we solving?             │
  │        │                                                    │
  │        ▼                                                    │
  │   2. COLLECT       Where does the data come from?           │
  │        │                                                    │
  │        ▼                                                    │
  │   3. CLEAN         Make messy data usable                   │
  │        │                                                    │
  │        ▼                                                    │
  │   4. EXPLORE       Understand what the data is telling us   │
  │        │                                                    │
  │        ▼                                                    │
  │   5. MODEL         Build a model or run analysis            │
  │        │                                                    │
  │        ▼                                                    │
  │   6. EVALUATE      Is the result actually good?             │
  │        │                                                    │
  │        ▼                                                    │
  │   7. DEPLOY        Share findings or ship the model         │
  │        │                                                    │
  │        └──────────────────────────────────────► (loop back) │
  │                                                             │
  └─────────────────────────────────────────────────────────────┘
```

Notice the arrow at the bottom — **it loops back**. Real projects are never linear. You explore the data and realise your question was wrong (back to Step 1). You build a model and find the data is too dirty (back to Step 3). You evaluate and discover the model is biased (back to Step 2 to collect better data).

This framework is based on **CRISP-DM** (Cross Industry Standard Process for Data Mining) — the most widely used data science process model in industry.

---
## Step 1 — Define the Problem

**The most important step. And the most skipped.**

Before touching any data, you must answer:

| Question | Why it matters |
|---|---|
| What is the business problem? | You need to know what success looks like |
| What decision will this analysis drive? | Otherwise why are you doing it? |
| Who is the audience? | A CEO wants a chart. An engineer wants an API. |
| What data do we have / need? | Determines if the project is even possible |
| Are there ethical concerns? | The retailer story above — ask this early |

### Turning a vague ask into a sharp question

| Vague Ask | Sharp Data Science Question |
|---|---|
| "Analyse our users" | "Which users are most likely to cancel their subscription in the next 30 days?" |
| "Look at sales data" | "Which product category drove the highest revenue growth in Q3 vs Q2?" |
| "Use AI for hiring" | "Which features of a resume correlate with 6-month performance ratings?" |
| "Check the data" | "Is there a statistically significant difference in exam scores between morning and afternoon batches?" |

**Rule of thumb:** If your question can't be answered with a number or a list, it's still too vague.

---
## Step 2 — Collect Data

Data comes from many places. A data scientist needs to know where to look.

### Sources of Data

```
┌─────────────────────────────────────────────────────────┐
│  WHERE DATA COMES FROM                                  │
│                                                         │
│  Internal                    External                   │
│  ─────────                   ────────                   │
│  • Company databases (SQL)   • Kaggle datasets          │
│  • CRM / ERP systems         • Government open data     │
│  • Log files                 • APIs (weather, maps)     │
│  • Spreadsheets              • Web scraping             │
│  • Survey responses          • Purchased datasets       │
│  • Sensor / IoT data         • Social media data        │
└─────────────────────────────────────────────────────────┘
```

### Key questions to ask about your data

- **How much data do I have?** (10 rows vs 10 million rows changes everything)
- **How old is this data?** (2019 sales data is useless for 2024 predictions)
- **Is it representative?** (A survey of only engineering students doesn't represent all students)
- **Are there privacy or legal concerns?** (Medical data, personal data — handle carefully)
- **Who collected it and how?** (Manual entry = more errors; sensors = different kinds of errors)

**In this course:** You'll collect data primarily from Kaggle and built-in sklearn datasets. In industry, Step 2 can take weeks or months.

---
## Step 3 — Clean the Data

> **"Data scientists spend 80% of their time cleaning data and 20% complaining about cleaning data."**  
> — A joke that's mostly true

Real data is almost never clean. You will encounter:

| Problem | Example |
|---|---|
| Missing values | Age column has blank cells |
| Wrong data types | Date stored as text: `"March 5"` instead of `2024-03-05` |
| Duplicates | Same customer appears twice with different spellings |
| Outliers | A salary column has a value of `₹9999999` (likely a data entry error) |
| Inconsistent categories | `"Male"`, `"male"`, `"M"`, `"m"` all mean the same thing |
| Wrong range | An age column has `"age = 250"` |

Cleaning decisions have consequences. If you drop every row with a missing value, you might remove 40% of your data. If you fill missing ages with the average, you're making an assumption that might be wrong.

**You'll go deep on this in Module 6.** For now, know that it exists and it's critical.

---
## Step 4 — Explore the Data (EDA)

**Exploratory Data Analysis (EDA)** is the detective phase. You're not building anything yet — you're investigating.

Questions you try to answer:
- What does the distribution of each column look like?
- Are there correlations between variables?
- What's surprising or unexpected?
- Does the data actually match the business question?

### Tools for EDA
- **Summary statistics** — mean, median, min, max, standard deviation
- **Visualisations** — histograms, box plots, scatter plots, correlation heatmaps
- **Cross-tabulation** — how does category A break down across category B?

EDA often **changes the question**. You come in asking "why did sales drop?" and discover during EDA that sales didn't actually drop — the reporting system had a bug. Now your question is "fix the reporting pipeline". 

**You'll spend all of Module 8 on EDA** — it's one of the most valuable skills you'll build.

---
## Step 5 — Model

This is the step everyone imagines when they think "data science" — but it's only one of seven.

"Model" can mean different things depending on the project:

| Type of project | What "model" means |
|---|---|
| Business analysis | A pivot table, a summary statistic, a trend line |
| Prediction | A machine learning model (linear regression, random forest, etc.) |
| Classification | A model that labels data (spam/not spam, fraud/not fraud) |
| Clustering | Grouping customers by behaviour — no labels needed |
| Reporting | A dashboard or automated report |

**Choosing the right model matters more than choosing the fanciest one.**  
A simple linear regression that's easy to explain often beats a complex neural network that nobody understands — especially when you have to justify a business decision.

---
## Step 6 — Evaluate

A model that isn't evaluated isn't a model — it's a guess with extra steps.

Evaluation means asking: **"Is this result actually good? Can we trust it?"**

### What evaluation looks like

| Task | Evaluation approach |
|---|---|
| Regression (predicting a number) | How far off are the predictions? (MAE, RMSE) |
| Classification (predicting a category) | How often is it right? (Accuracy, F1 Score) |
| Clustering | Do the groups make sense to a human expert? |
| Analysis / Dashboard | Does the finding match known reality? Did stakeholders validate it? |

### The biggest evaluation trap — overfitting

A model that memorises the training data scores 100% on training data but fails on new data. You evaluate on **data the model has never seen** (the test set) to catch this.

**You'll learn all the evaluation metrics in Modules 11 and 12.**

---
## Step 7 — Deploy / Communicate

A model that lives only in a Jupyter notebook helps nobody.

"Deployment" means getting the work into the hands of the people who need it. That can mean very different things:

| Audience | Deployment format |
|---|---|
| Business team / management | PowerPoint / PDF report with charts |
| Other developers | An API that returns predictions (`/predict?customer_id=123`) |
| End users | A web app or dashboard (Streamlit, Tableau, Power BI) |
| Automated pipeline | A scheduled script that runs daily and sends a report |

### Communication is the hidden skill

Knowing *how* to present findings is as important as finding them. A chart with no title, no axis labels, and no takeaway statement is worse than no chart at all — it wastes your audience's time and makes your work look sloppy.

**In this course:** Every project ends with a section called "Insights and Recommendations" — you'll practice communicating findings clearly every single module.

---
## 2. The Lifecycle Is a Loop

Here's what really happens on a real project:

```
  Day 1:  Define → "Predict which students will fail the exam"
  Day 2:  Collect → Get 3 years of student records
  Day 3:  Clean → Discover 40% of attendance data is missing
            ↑
            └── Back to Collect — can we get the attendance from another source?

  Day 5:  Clean → Fixed. Now exploring data.
  Day 6:  Explore → Realise exam scores are highly correlated with assignment grades
            ↑
            └── Refine the question: "Can we predict failure from Week 4 assignment grades alone?"
                 (Back to Step 1 — better question!)

  Day 8:  Model → Build logistic regression
  Day 9:  Evaluate → Accuracy is 91% but it misses 60% of actual failures
            ↑
            └── Back to Model — wrong metric, wrong model. Rebuild.

  Day 11: Evaluate → Better. 85% accuracy, catches 80% of failures.
  Day 12: Deploy → Email report to teachers every Monday
```

**The loop is not a sign of failure — it's the process working correctly.**  
Projects that never loop back are usually the ones that deliver wrong answers with high confidence.

## 3. Mini Project Walkthrough — Student Marks Analysis

Let's walk through all 7 steps on a small, real example.  
We'll use only Python built-ins — no Pandas, no NumPy yet.  
This is your preview of what you'll do properly in Module 3.

**The scenario:** A teacher collected marks from 15 students across 3 subjects. She wants to know: *"Which students are at risk of failing, and which subject needs the most attention?"*

````python
# ── STEP 1: Define ───────────────────────────────────────────────────────────
# Question: Which students are at risk? Which subject needs attention?
# Threshold: < 40 marks in any subject = at risk
# Output: List of at-risk students + weakest subject

print("Project: Student Performance Analysis")
print("Question: Who is at risk of failing, and which subject is weakest?")
print("Audience: Class teacher")
print("Success: A clear list + one actionable recommendation")
````

````python
# ── STEP 2: Collect ──────────────────────────────────────────────────────────
# In a real project this would come from a CSV file or a school database.
# Here we define it directly — same concept.

students = [
    # [name,       maths, science, english]
    ["Aarav",      78,    82,      None],   # missing english score
    ["Priya",      91,    88,      95],
    ["Rohan",      35,    42,      38],     # struggling
    ["Simran",     67,    71,      74],
    ["Dev",        55,    None,    60],     # missing science score
    ["Meera",      88,    90,      85],
    ["Karan",      22,    30,      28],     # severely struggling
    ["Ananya",     74,    68,      72],
    ["Vihaan",     60,    55,      62],
    ["Ishaan",     45,    38,      41],
    ["Riya",       83,    79,      80],
    ["Arjun",      92,    95,      89],
    ["Kavya",      38,    44,      35],     # struggling
    ["Nikhil",     70,    66,      75],
    ["Saanvi",     55,    61,      None],   # missing english score
]

print(f"Collected data for {len(students)} students.")
print(f"Subjects: Maths, Science, English")
print(f"\nFirst 3 rows:")
print(f"{'Name':<10} {'Maths':>7} {'Science':>9} {'English':>9}")
print("-" * 40)
for row in students[:3]:
    name, maths, science, english = row
    print(f"{name:<10} {str(maths):>7} {str(science):>9} {str(english):>9}")
````

````python
# ── STEP 3: Clean ────────────────────────────────────────────────────────────
# Problem found: 3 students have missing (None) scores.
# Decision: replace None with the subject average for that subject.
# Reason: we don't want to remove those students entirely.

def subject_average(students, subject_index):
    """Calculate mean ignoring None values."""
    scores = [row[subject_index] for row in students if row[subject_index] is not None]
    return round(sum(scores) / len(scores), 1)

maths_avg   = subject_average(students, 1)
science_avg = subject_average(students, 2)
english_avg = subject_average(students, 3)

print(f"Subject averages (used to fill missing values):")
print(f"  Maths:   {maths_avg}")
print(f"  Science: {science_avg}")
print(f"  English: {english_avg}")

# Fill in missing values
clean_students = []
for row in students:
    name    = row[0]
    maths   = row[1] if row[1] is not None else maths_avg
    science = row[2] if row[2] is not None else science_avg
    english = row[3] if row[3] is not None else english_avg
    clean_students.append([name, maths, science, english])

missing_before = sum(1 for row in students for val in row[1:] if val is None)
print(f"\nMissing values filled: {missing_before} cells imputed with subject average.")
print("Data is now clean.")
````

````python
# ── STEP 4: Explore ──────────────────────────────────────────────────────────
# Now we look at the data — distributions, patterns, anything surprising.

import statistics

subjects = ["Maths", "Science", "English"]

print(f"{'Subject':<10} {'Min':>5} {'Max':>5} {'Mean':>7} {'Median':>8} {'Std Dev':>9}")
print("-" * 48)

for i, subject in enumerate(subjects, start=1):
    scores = [row[i] for row in clean_students]
    print(
        f"{subject:<10}"
        f"{min(scores):>5}"
        f"{max(scores):>5}"
        f"{statistics.mean(scores):>7.1f}"
        f"{statistics.median(scores):>8.1f}"
        f"{statistics.stdev(scores):>9.1f}"
    )

print()

# Distribution: how many students are in each grade band?
def grade(score):
    if score >= 80: return "A (≥80)"
    if score >= 60: return "B (60-79)"
    if score >= 40: return "C (40-59)"
    return "F (<40)"

print("Grade distribution across all subjects:")
all_scores = [row[i] for row in clean_students for i in range(1, 4)]
grade_counts = {"A (≥80)": 0, "B (60-79)": 0, "C (40-59)": 0, "F (<40)": 0}
for s in all_scores:
    grade_counts[grade(s)] += 1

total_scores = len(all_scores)
for g, count in grade_counts.items():
    bar = "█" * count
    print(f"  {g:<12} {bar} {count} ({count/total_scores*100:.0f}%)")
````

````python
# ── STEP 5: Model / Analyse ──────────────────────────────────────────────────
# No ML needed here — our "model" is a clear definition:
# At-risk = scored below 40 in ANY subject
# Weakest subject = lowest average score

FAIL_THRESHOLD = 40

at_risk = []
for row in clean_students:
    name   = row[0]
    scores = row[1:]
    failed_subjects = [
        subjects[i] for i, s in enumerate(scores) if s < FAIL_THRESHOLD
    ]
    if failed_subjects:
        avg = round(sum(scores) / len(scores), 1)
        at_risk.append((name, avg, failed_subjects))

# Which subject has the lowest average?
subject_avgs = {
    subjects[i]: round(statistics.mean([row[i+1] for row in clean_students]), 1)
    for i in range(3)
}
weakest_subject = min(subject_avgs, key=subject_avgs.get)

print("Analysis complete.")
print(f"At-risk students found  : {len(at_risk)}")
print(f"Weakest subject         : {weakest_subject} (avg: {subject_avgs[weakest_subject]})")
````

````python
# ── STEP 6: Evaluate ─────────────────────────────────────────────────────────
# Sanity-check: Do the at-risk students look right?
# Does the weakest subject make sense given what we saw in the EDA?

print("Results look reasonable?")
print()

# Check: does weakest subject match our EDA minimum?
eda_minimums = {
    subjects[i]: min(row[i+1] for row in clean_students)
    for i in range(3)
}
print(f"Subject minimums from EDA : {eda_minimums}")
print(f"Subject averages          : {subject_avgs}")
print(f"Weakest subject (by avg)  : {weakest_subject}  ← consistent with EDA")
print()

# Check: are the at-risk students actually below 40?
print("At-risk validation:")
for name, avg, failed in at_risk:
    print(f"  {name:<10} overall avg: {avg}  failed in: {', '.join(failed)}")
````

````python
# ── STEP 7: Communicate ──────────────────────────────────────────────────────
# Present the findings clearly for the teacher — the actual audience.

print("=" * 55)
print("     STUDENT PERFORMANCE REPORT — CLASS SUMMARY")
print("=" * 55)
print()
print(f"Total students assessed : {len(clean_students)}")
print(f"Students at risk        : {len(at_risk)} ({len(at_risk)/len(clean_students)*100:.0f}%)")
print()

print("AT-RISK STUDENTS (scored below 40 in at least one subject):")
print(f"  {'Name':<10} {'Overall Avg':>12}  {'Needs help in'}")
print("  " + "-" * 48)
for name, avg, failed in sorted(at_risk, key=lambda x: x[1]):
    print(f"  {name:<10} {avg:>12}  {', '.join(failed)}")

print()
print(f"SUBJECT PERFORMANCE:")
for subj, avg in sorted(subject_avgs.items(), key=lambda x: x[1]):
    bar = "█" * int(avg / 5)
    print(f"  {subj:<10} {bar} {avg}")

print()
print("RECOMMENDATION:")
print(f"  → {weakest_subject} has the lowest class average ({subject_avgs[weakest_subject]}).")
print(f"    Consider dedicating an extra revision session before the next test.")
print(f"  → {len(at_risk)} students need immediate intervention. Suggest one-on-one review.")
print()
print("=" * 55)
````

**That was a complete Data Science project.**

No machine learning. No fancy libraries. Just a clear question, real data, careful cleaning, honest analysis, and a recommendation the teacher can actually act on.

## 4. Where Projects Fail — The Most Common Mistakes

| Step | Common failure |
|---|---|
| Define | Question is too vague; no one agrees on what success means |
| Collect | Using biased, outdated, or unrepresentative data |
| Clean | Dropping too many rows; not documenting what was changed |
| Explore | Skipping EDA entirely and jumping straight to modelling |
| Model | Choosing the most complex model instead of the most appropriate one |
| Evaluate | Only evaluating on training data (overfitting trap) |
| Deploy | Building something the audience can't understand or won't use |

The most dangerous failure is at **Step 1** — if the question is wrong, it doesn't matter how well you execute every other step.

---
## ⚠️ Common Misconceptions

### Misconception 1: "The lifecycle is a checklist you go through once"
**Wrong.** It's a loop. Real projects go back and forth between steps constantly. Discovering something unexpected in EDA often means redefining the question. A poor evaluation score means going back to model — or further back to clean.

### Misconception 2: "Step 5 (Model) is the most important step"
**Wrong.** Step 1 (Define) and Step 3 (Clean) have the most impact on project success. A brilliantly trained model on dirty data gives you precise wrong answers. A simple model on clean, well-understood data gives you reliable, useful answers.

### Misconception 3: "Deploy means putting it on the internet"
**Not always.** For most data science projects at this level, "deploy" means presenting findings clearly to someone who can act on them — a teacher, a manager, a team. A well-crafted PDF report or a clear Jupyter notebook summary counts as deployment.

## 📌 Key Takeaways

- **The lifecycle has 7 steps: Define → Collect → Clean → Explore → Model → Evaluate → Deploy.** Every real project follows this — whether it takes 1 day or 6 months. Skipping steps doesn't save time; it creates problems you'll have to fix later.

- **It's a loop, not a straight line.** Discovering something unexpected at any step can — and should — send you back to an earlier one. This is the process working correctly, not failing.

- **Step 1 (Define) has the biggest impact.** A wrong question means all the work that follows is wasted. Before touching data, make sure you know exactly what decision this analysis will drive and who will act on it.
