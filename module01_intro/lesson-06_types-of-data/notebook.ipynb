## 1. The Full Taxonomy

Every piece of data you will ever work with fits somewhere in this tree:

```
DATA
│
├── STRUCTURED  (organised into rows and columns)
│   │
│   ├── QUANTITATIVE  (numbers — can do arithmetic on them)
│   │   ├── Continuous   (any value in a range: height, salary, temperature)
│   │   └── Discrete     (only whole counts: students, goals, orders)
│   │
│   └── QUALITATIVE  (categories — cannot do arithmetic on them)
│       ├── Nominal      (no natural order: city, gender, colour, blood type)
│       └── Ordinal      (has natural order: rating 1–5, education, t-shirt size)
│
├── SEMI-STRUCTURED  (has some structure but not a rigid table)
│   └── JSON, XML, CSV with inconsistent schema, log files
│
└── UNSTRUCTURED  (no predefined format)
    └── Text, Images, Audio, Video
```

Most of this course works with **structured data** — tables with rows and columns.  
But you need to know where other types live so you understand the full landscape.

## 2. Structured vs Semi-Structured vs Unstructured

### Structured Data
Lives in a table. Every row has the same columns. You can open it in Excel or query it with SQL.

```
| student_id | name    | age | score | grade |
|------------|---------|-----|-------|-------|
| 1          | Priya   | 20  | 88    | A     |
| 2          | Rohan   | 21  | 62    | B     |
| 3          | Meera   | 19  | 45    | C     |
```

**Examples:** Bank transactions, hospital records, IPL scorecards, attendance sheets, e-commerce orders.

---

### Semi-Structured Data
Has *some* organisation — keys, tags, markers — but not a rigid table structure. Each record may have different fields.

```json
{ "order_id": 101, "customer": "Priya",
  "items": [{"name": "Earbuds", "qty": 1}, {"name": "Case", "qty": 2}],
  "address": {"city": "Mumbai", "pin": "400001"} }
```

**Examples:** JSON from APIs, XML configuration files, server log files, HTML web pages.  
You'll work with JSON in the Bonus module on APIs.

---

### Unstructured Data
No predefined format. The most common type of data in the world — and the hardest to analyse without deep learning.

| Type | Examples | Typical tool |
|---|---|---|
| Text | Reviews, tweets, news articles, emails | NLP, transformers |
| Images | X-rays, product photos, satellite images | Computer vision, CNNs |
| Audio | Call recordings, music, speech | Wav2Vec, Whisper |
| Video | CCTV footage, YouTube videos | Video CNNs, multimodal models |

**In this course:** We work exclusively with structured data. Unstructured data requires deep learning — a separate course.

## 3. Quantitative Data — Numbers You Can Calculate With

Quantitative data represents measurable quantities. You can add, subtract, multiply, and divide them meaningfully.

### 3.1 Continuous Data

> **Can take any value within a range — including decimals.**

The key test: *Can there be a meaningful value between any two values?*  
Between 1.5 kg and 1.6 kg, there is 1.55 kg. Between 30°C and 31°C, there is 30.7°C. → **Continuous.**

| Column | Values | Why continuous |
|---|---|---|
| Height | 162.5 cm, 178.3 cm | Any value possible between measurements |
| Salary | ₹45,000.50 | Can be any amount |
| Temperature | 36.6°C | Decimal values are meaningful |
| Time taken | 23.7 seconds | Sub-second values exist |
| Blood pressure | 120/80 mmHg | Measured on a continuous scale |

**Right charts:** Histogram, box plot, line plot, scatter plot  
**Right statistics:** Mean, median, standard deviation, percentiles

---

### 3.2 Discrete Data

> **Can only take specific whole values — usually counts.**

The key test: *Does a fractional value make no sense?*  
You can't have 2.7 students or 4.3 goals. → **Discrete.**

| Column | Values | Why discrete |
|---|---|---|
| Number of students | 35, 38, 42 | Can't have 35.7 students |
| Goals scored | 0, 1, 2, 3 | Must be whole numbers |
| Number of orders | 150, 203 | Count of items |
| Age (in years) | 18, 19, 20 | Measured in whole years |
| Number of rooms | 1, 2, 3 | Can't have 2.5 rooms |

**Right charts:** Bar chart, histogram (with integer bins), scatter plot  
**Right statistics:** Count, mode, median (mean is usable but note it may not be a whole number)

**Note:** Age is technically continuous (you are 20 years and 147 days old), but when recorded as whole years it becomes discrete. Context determines classification.

---
## 4. Qualitative Data — Categories You Cannot Calculate With

Qualitative data describes characteristics or groups. You cannot meaningfully add or average them.

### 4.1 Nominal Data

> **Categories with no natural order — swapping them changes nothing.**

The key test: *Is there a meaningful "greater than" relationship between categories?*  
Is Mumbai > Delhi? Is Red > Blue? No. → **Nominal.**

| Column | Values | Why nominal |
|---|---|---|
| City | Mumbai, Delhi, Chennai | No city is "more than" another |
| Blood type | A, B, AB, O | No natural ordering |
| Product category | Electronics, Clothing | No rank between them |
| Gender | Male, Female, Other | No meaningful order |
| Team name | MI, CSK, RCB | No ranking in the name itself |

**Right charts:** Bar chart, pie chart, countplot  
**Right statistics:** Frequency, mode, count  
**Wrong:** Mean (`mean(["Mumbai", "Delhi"])` — nonsense)

---

### 4.2 Ordinal Data

> **Categories with a meaningful order — but the gaps between them are not necessarily equal.**

The key test: *Can I say one value is clearly "more" or "higher" than another?*  
"Excellent" > "Good" > "Poor" — yes. → **Ordinal.**

But: the gap between "Poor" and "Good" is not necessarily the same size as the gap between "Good" and "Excellent". That's what makes it ordinal, not continuous.

| Column | Values | Why ordinal |
|---|---|---|
| Rating | 1★, 2★, 3★, 4★, 5★ | Clear order, but gap between 1 and 2 may differ from gap between 4 and 5 |
| Education level | School → UG → PG → PhD | Clear hierarchy |
| T-shirt size | S, M, L, XL | Ordered by size |
| Pain scale | None, Mild, Moderate, Severe | Clear order |
| Customer satisfaction | Dissatisfied → Neutral → Satisfied | Ordered |

**Right charts:** Bar chart (ordered), countplot  
**Right statistics:** Median, mode, frequency — mean is debatable (often computed in practice but technically imprecise)  
**ML treatment:** Encode with numbers that preserve the order (1, 2, 3, 4, 5) — covered in Module 6

## 5. Quick Reference — The Decision Tree

When you encounter any column, ask these questions in order:

```
Is the value a NUMBER?
│
├── YES → Can a meaningful decimal exist between any two values?
│         ├── YES → CONTINUOUS  (height, salary, temperature)
│         └── NO  → DISCRETE    (student count, goals, number of items)
│
└── NO  → It's a CATEGORY. Does the category have a natural order?
          ├── NO  → NOMINAL     (city, blood type, colour, team name)
          └── YES → ORDINAL     (rating, size, education level, satisfaction)
```

**The tricky case — numbers stored as categories:**  
A "pin code" column contains numbers (400001, 110001) but it's **nominal** — you can't say pin code 400001 is "greater than" 110001 in any meaningful sense, and averaging pin codes is nonsense.

## 6. Code Demo — Classifying a Real Dataset's Columns

````python
# A realistic student dataset — each row is one student
# We'll classify every column by data type

dataset = [
    # student_id, name,      city,       age, height_cm, score, grade,  satisfied, num_absent
    [101, "Priya",   "Mumbai",    20, 162.5, 88, "A",  "Very Satisfied", 2],
    [102, "Rohan",   "Delhi",     21, 175.0, 62, "B",  "Neutral",        5],
    [103, "Meera",   "Mumbai",    19, 158.3, 45, "C",  "Dissatisfied",   9],
    [104, "Karan",   "Bengaluru", 22, 180.2, 91, "A",  "Very Satisfied", 0],
    [105, "Ananya",  "Chennai",   20, 163.0, 74, "B",  "Satisfied",      3],
    [106, "Vihaan",  "Delhi",     21, 172.5, 55, "C",  "Neutral",        7],
    [107, "Ishaan",  "Pune",      23, 168.8, 38, "F",  "Dissatisfied",  12],
    [108, "Riya",    "Mumbai",    19, 155.5, 83, "A",  "Satisfied",      1],
]

columns = [
    "student_id", "name", "city", "age",
    "height_cm", "score", "grade", "satisfaction", "num_absences"
]

# Classification of each column
column_types = {
    "student_id"    : ("Nominal",    "Qualitative", "It's an ID — no arithmetic makes sense"),
    "name"          : ("Nominal",    "Qualitative", "Names have no order or arithmetic"),
    "city"          : ("Nominal",    "Qualitative", "No city is 'greater than' another"),
    "age"           : ("Discrete",   "Quantitative","Recorded in whole years — no 20.7 years"),
    "height_cm"     : ("Continuous", "Quantitative","162.5 cm — decimal values are meaningful"),
    "score"         : ("Continuous", "Quantitative","Could be 88.5; treated as continuous"),
    "grade"         : ("Ordinal",    "Qualitative", "A > B > C > F — clear order, no arithmetic"),
    "satisfaction"  : ("Ordinal",    "Qualitative", "Very Satisfied > Satisfied > Neutral — ordered"),
    "num_absences"  : ("Discrete",   "Quantitative","Count of days — must be whole number"),
}

print(f"{'Column':<16} {'Type':<12} {'Category':<14} {'Reason'}")
print("─" * 80)
for col in columns:
    dtype, category, reason = column_types[col]
    print(f"{col:<16} {dtype:<12} {category:<14} {reason}")
````

**Notice `student_id`:** It looks like a number but it's Nominal. Averaging student IDs (101+102+103)/3 = 102 — that result is meaningless. Always ask *"does arithmetic on this column produce a useful result?"* before treating a number as quantitative.

## 7. What Goes Wrong With the Wrong Type — Live Errors

````python
# Pulling column values from our dataset
row_index = {col: i for i, col in enumerate(columns)}

def get_column(col_name):
    idx = row_index[col_name]
    return [row[idx] for row in dataset]

cities       = get_column("city")
heights      = get_column("height_cm")
grades       = get_column("grade")
scores       = get_column("score")
student_ids  = get_column("student_id")

print("Cities   :", cities)
print("Heights  :", heights)
print("Grades   :", grades)
print("Scores   :", scores)
print("IDs      :", student_ids)
````

````python
import statistics

# ✅ CORRECT — mean of a continuous column
mean_height = statistics.mean(heights)
mean_score  = statistics.mean(scores)
print(f"Mean height : {mean_height:.1f} cm  ← makes sense")
print(f"Mean score  : {mean_score:.1f}     ← makes sense")
````

````python
# ❌ WRONG — mean of a nominal column (Python won't stop you — but the result is nonsense)
# This will raise a TypeError because strings can't be averaged

try:
    mean_city = statistics.mean(cities)
    print(f"Mean city: {mean_city}")
except TypeError as e:
    print(f"TypeError on nominal data: {e}")
    print("→ You CANNOT compute the mean of a city column.")
    print("→ Use: mode (most common city) or value_counts instead.")
````

````python
# ❌ SILENT WRONG — mean of student_id (Python happily computes it, but it's meaningless)
mean_id = statistics.mean(student_ids)
print(f"Mean student_id: {mean_id}")
print("↑ Python computed this with no error.")
print("  But what does 'average student ID = 104.5' mean? Nothing.")
print("  This is the dangerous kind of mistake — no error, just a wrong answer.")
````

````python
# ❌ WRONG — sorting ordinal data without specifying the order
# Python sorts grades alphabetically: A, B, C, F — which happens to be correct here.
# But satisfaction sorts WRONG:

satisfaction = get_column("satisfaction")
print("Satisfaction values:", satisfaction)
print()

print("Alphabetical sort (WRONG for ordinal):")
print(sorted(satisfaction))
print("↑ 'Dissatisfied' comes first alphabetically — but it should come last!")
print()

# ✅ CORRECT — define the order explicitly
order = ["Dissatisfied", "Neutral", "Satisfied", "Very Satisfied"]
print("Correct ordinal sort:")
print(sorted(satisfaction, key=lambda x: order.index(x)))
print("↑ Now 'Dissatisfied' is first (lowest) and 'Very Satisfied' is last (highest).")
````

**The three failure modes:**

| Mistake | What happens | How to detect |
|---|---|---|
| Mean of a text nominal column | `TypeError` — obvious | Error message |
| Mean of a numeric nominal column (like ID) | Computes fine, means nothing | Domain knowledge — ask "does this result make sense?" |
| Sorting ordinal without specifying order | Wrong order, no error | Know your data — define the order explicitly |

## 8. Right Chart for Each Data Type

Chart choice is the first place data type knowledge gets applied. You'll practise this in Module 7 — but knowing the rules now prevents bad visualisations from the start.

```
DATA TYPE          GOOD CHARTS                   AVOID
─────────────────  ────────────────────────────  ──────────────────────────
Continuous         Histogram, Box Plot,           Pie chart (too many slices)
                   Line plot, Scatter plot

Discrete           Bar chart, Histogram           Pie chart for large counts
                   (integer bins)

Nominal            Bar chart, Pie chart           Line plot (implies order),
                   Countplot                       Histogram

Ordinal            Bar chart (ordered),           Pie chart, unordered bar
                   Countplot

Two Continuous     Scatter plot, Line plot        Bar chart

Continuous vs      Box plot, Violin plot,         Line plot
Categorical        Grouped bar chart
```

## 10. Python Data Types vs Statistical Data Types

Python has its own type system. It doesn't map perfectly to statistical types — and that's where confusion creeps in.

````python
# Python types vs statistical types

examples = [
    # (value,          python_type,   statistical_type,  note)
    (162.5,            "float",       "Continuous",       "Straightforward match"),
    (35,               "int",         "Discrete",         "Count of students"),
    (101,              "int",         "Nominal",          "Student ID — number but not quantitative"),
    ("Mumbai",         "str",         "Nominal",          "Text category"),
    ("Very Satisfied", "str",         "Ordinal",          "Text with hidden order"),
    ("A",              "str",         "Ordinal",          "Grade — ordered A > B > C"),
    (True,             "bool",        "Nominal (binary)", "Yes/No, 0/1 — two categories"),
    ("2024-03-15",     "str/date",    "Continuous",       "Time is continuous; special handling needed"),
    (400001,           "int",         "Nominal",          "Pin code — integer but nominal"),
]

print(f"{'Value':<18} {'Python type':<14} {'Statistical type':<20} {'Note'}")
print("─" * 78)
for val, py_type, stat_type, note in examples:
    print(f"{str(val):<18} {py_type:<14} {stat_type:<20} {note}")
````

````python
# Python's type() tells you the Python type — NOT the statistical type.
# You need domain knowledge to determine the statistical type.

values = [
    (162.5,  "height_cm"),
    (101,    "student_id"),
    (400001, "pin_code"),
    (35,     "num_students"),
]

print("Python type alone is NOT enough to determine statistical type:")
print()
print(f"{'Column':<15} {'Value':<10} {'Python type':<10} {'Statistical type':<18} {'Correct operation'}")
print("─" * 80)

stat_types = ["Continuous", "Nominal", "Nominal", "Discrete"]
operations = ["mean, std", "mode only", "mode only", "sum, mean"]

for (val, col), stat_type, op in zip(values, stat_types, operations):
    print(f"{col:<15} {val:<10} {type(val).__name__:<10} {stat_type:<18} {op}")
````

**The lesson:** `type(x)` tells you how Python stores the value. It does NOT tell you what operations are statistically valid. You need to understand your data domain to make that call.

## 11. Special Case — Time Series Data

Time series data is a sequence of values recorded at regular time intervals. It's technically continuous, but it has special properties that require different techniques.

```
Date          Temperature
────────────  ───────────
2024-01-01    18.5°C
2024-01-02    19.2°C
2024-01-03    17.8°C
...
```

**What makes it special:**
- Order matters — January 1 always comes before January 2
- Values are often correlated with their neighbours (today's temp is close to yesterday's)
- Patterns repeat seasonally (sales spike every Diwali)

**Where you see it:** Stock prices, weather data, app usage over time, sales by day, heart rate monitor

## ⚠️ Common Misconceptions

### Misconception 1: "If it's stored as a number, it's quantitative"
**Wrong.** Pin codes, student IDs, phone numbers, and zip codes are all stored as integers — but they're nominal. The test is never *how it's stored*; it's *whether arithmetic on it produces a meaningful result.*

### Misconception 2: "Ordinal data can always be treated as continuous"
**Partially wrong.** Many analysts compute the mean of a 1–5 rating scale, and in practice this is often useful. But strictly speaking, the gap between 1 and 2 may not equal the gap between 4 and 5. A customer who gave 1 star might be furious (a very different experience from a 2-star), while 4 and 5 might be very similar. Be careful about interpreting the mean of ordinal data as though it were a precise measurement.

### Misconception 3: "Unstructured data is too hard to work with"
**Becoming less true.** With modern NLP and computer vision tools, working with text and images is increasingly accessible. But for this course, structured data will take you to 90% of real-world data science problems — master it first.

## 📌 Key Takeaways

- **Every column fits the taxonomy: Continuous · Discrete · Nominal · Ordinal.** Know which one before touching any column. This one decision determines which charts, which statistics, and which ML preprocessing steps are valid.

- **Python's type is not the same as the statistical type.** A number stored as `int` can be nominal (student ID, pin code). A string can be ordinal ("S", "M", "L"). Always apply domain knowledge — the computer won't stop you from doing nonsense arithmetic on a nominal column.

- **The full data landscape: Structured → Semi-structured → Unstructured.** This course focuses on structured tabular data, which covers 90% of real business problems. Unstructured data (images, text) requires deep learning — a different course built on top of this one.
