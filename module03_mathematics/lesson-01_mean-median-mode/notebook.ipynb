---

> Before any ML algorithm runs, before any chart is drawn, a data scientist asks: *What is the typical value?* These three numbers answer that question.

Imagine you're a placement officer at an engineering college. A company asks: "What is the typical salary your graduates get?" You have 50 offer letters. Which number do you report?

- If most students got ₹6–8 LPA but one got ₹1.2 Crore from Google, the **mean** (average) would be inflated — misleading.
- The **median** (middle value) would tell the company what a *typical* student earns.
- The **mode** (most common value) would tell them what package appears most often on offer letters.

**Mean**, **median**, and **mode** are called *measures of central tendency* — they each describe where the "centre" of your data is, from three different angles. Knowing when to use which one is one of the first real skills of a data scientist.

## 1. Mean (Arithmetic Average)

### Definition

$$\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n} = \frac{x_1 + x_2 + \cdots + x_n}{n}$$

The mean is the **sum of all values divided by the count**. Every value contributes equally to the result. This is its strength — and its weakness.

````python
def mean(data):
    """Compute the arithmetic mean of a non-empty list of numbers."""
    if not data:
        raise ValueError("Cannot compute mean of an empty list")
    return sum(data) / len(data)


# Exam scores for a class of 10 students
scores = [72, 85, 68, 91, 76, 88, 55, 79, 83, 67]

print(f"Scores  : {scores}")
print(f"Count   : {len(scores)}")
print(f"Sum     : {sum(scores)}")
print(f"Mean    : {mean(scores):.1f}")
````

### 1.1 The Outlier Problem

An **outlier** is a value far from the rest. The mean treats every value equally — so one extreme value can pull the mean far from where most data sits.

````python
# Monthly salaries (in ₹) at a small startup — 9 employees + 1 founder
salaries = [35000, 42000, 38000, 45000, 40000,
            37000, 43000, 39000, 41000, 1_200_000]  # founder's salary

mean_with    = mean(salaries)
mean_without = mean(salaries[:-1])   # remove the founder

print(f"All 10 salaries  : mean = ₹{mean_with:,.0f}")
print(f"9 employees only : mean = ₹{mean_without:,.0f}")
print()
print("The founder's ₹12 lakh salary pulls the mean up by")
print(f"₹{mean_with - mean_without:,.0f} — making the company look much more")
print("generous than it actually is for typical employees.")
print()
print("This is why median is preferred for income/salary data.")
````

### 1.2 Weighted Mean

In a standard mean, every value counts equally. In a **weighted mean**, some values count more than others.

$$\bar{x}_w = \frac{\sum_{i=1}^{n} w_i \cdot x_i}{\sum_{i=1}^{n} w_i}$$

**Real-world use:** GPA calculation, final exam scores where assignments and finals have different weightages.

````python
def weighted_mean(values, weights):
    """
    Compute the weighted mean.
    values  : list of numeric scores
    weights : list of weights (same length, must sum to > 0)
    """
    if len(values) != len(weights):
        raise ValueError("values and weights must have the same length")
    total_weight = sum(weights)
    if total_weight == 0:
        raise ValueError("Weights must not all be zero")
    return sum(v * w for v, w in zip(values, weights)) / total_weight


# Priya's semester evaluation
components = ["Assignments", "Mid-sem",  "Practicals", "End-sem"]
scores_pct = [          82,         74,            90,        68]   # scores out of 100
weightages = [          20,         20,            10,        50]   # percentage weight

print("Priya's Semester Marks")
print(f"{'Component':<14} {'Score':>7} {'Weight':>8} {'Contribution':>14}")
print("─" * 47)
for comp, score, wt in zip(components, scores_pct, weightages):
    contribution = score * wt / 100
    print(f"  {comp:<12} {score:>7}   {wt:>5}%   {contribution:>12.1f}")
print("─" * 47)

simple_avg   = mean(scores_pct)
weighted_avg = weighted_mean(scores_pct, weightages)

print(f"\n  Simple mean    : {simple_avg:.1f}")
print(f"  Weighted mean  : {weighted_avg:.1f}  ← the real final score")
print()
print("Because the end-sem exam has 50% weight, a lower score")
print("there pulls the final result down more than assignments.")
````

## 2. Median

### Definition

The **median** is the *middle value* when the data is sorted. Half the values lie below it, half above.

- If `n` is **odd**: median = value at position `n // 2` (0-indexed) in the sorted list
- If `n` is **even**: median = average of the two middle values

The key insight: **the median ignores outliers completely**. It only cares about which value sits in the middle, not how extreme the other values are.

````python
def median(data):
    """Compute the median of a non-empty list of numbers."""
    if not data:
        raise ValueError("Cannot compute median of an empty list")
    sorted_data = sorted(data)    # don't modify the original
    n   = len(sorted_data)
    mid = n // 2
    if n % 2 == 1:                # odd count: exact middle
        return sorted_data[mid]
    else:                         # even count: average two middles
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2


# Odd count
odd_data = [15, 22, 8, 31, 19]   # sorted: [8, 15, 19, 22, 31]
print(f"Odd  data   : {odd_data}")
print(f"Sorted      : {sorted(odd_data)}")
print(f"Median      : {median(odd_data)}  (middle value)")

print()

# Even count
even_data = [15, 22, 8, 31, 19, 25]  # sorted: [8, 15, 19, 22, 25, 31]
print(f"Even data   : {even_data}")
print(f"Sorted      : {sorted(even_data)}")
print(f"Middle pair : (19 + 22) / 2 = {(19+22)/2}")
print(f"Median      : {median(even_data)}")
````

````python
# The median's superpower: robustness to outliers

# Same salary dataset from before
salaries = [35000, 42000, 38000, 45000, 40000,
            37000, 43000, 39000, 41000, 1_200_000]

print("Impact of the ₹12 lakh outlier:")
print(f"\n  {'Measure':<10} {'With outlier':>16} {'Without outlier':>17}")
print("  " + "─" * 46)
print(f"  {'Mean':<10} ₹{mean(salaries):>14,.0f} ₹{mean(salaries[:-1]):>15,.0f}")
print(f"  {'Median':<10} ₹{median(salaries):>14,.0f} ₹{median(salaries[:-1]):>15,.0f}")
print()
print("The mean swings by ₹1.3 lakh when the outlier is removed.")
print("The median barely moves — it stays close to the typical employee salary.")
````

## 3. Mode

### Definition

The **mode** is the value that appears **most frequently** in the dataset. Unlike mean and median, mode works for both numbers *and* categories.

- A dataset can have **one mode** (unimodal), **two modes** (bimodal), or **no mode** (all values appear once).

````python
def mode(data):
    """
    Return the mode(s) of data.
    If multiple values tie for the highest frequency, returns all of them.
    Returns an empty list if data is empty.
    """
    if not data:
        return []

    # Build frequency count
    freq = {}
    for item in data:
        freq[item] = freq.get(item, 0) + 1

    max_count = max(freq.values())

    # All values that appear max_count times
    modes = sorted(k for k, v in freq.items() if v == max_count)
    return modes


# Unimodal — one clear winner
grades = ["B", "A", "C", "B", "B", "A", "C", "B", "A", "B"]
print(f"Grades     : {grades}")
print(f"Mode       : {mode(grades)}  ← B appears most (5 times)")

# Bimodal — two tied modes
scores = [70, 85, 70, 90, 85, 75, 60, 85, 70, 80]
print(f"\nScores     : {scores}")
print(f"Mode       : {mode(scores)}  ← 70 and 85 both appear 3 times")

# No mode — all values unique
heights = [165, 170, 172, 168, 175]
print(f"\nHeights    : {heights}")
print(f"Mode       : {mode(heights)}  ← all appear once, all are modes")
````

````python
# Mode's true power: categorical data
# Mean and median can't handle categories — mode can

departments = ["CS", "ME", "CS", "EE", "CS", "IT", "ME", "CS",
               "EE", "IT", "CS", "ME", "CS", "IT", "EE"]
cities = ["Mumbai", "Delhi", "Pune", "Mumbai", "Bangalore", "Mumbai",
          "Delhi", "Pune", "Mumbai", "Chennai", "Mumbai", "Delhi"]

print(f"Most common department : {mode(departments)}")
print(f"Most common city       : {mode(cities)}")
print()

# E-commerce / product analytics: most sold size
shirt_sizes = ["M", "L", "S", "M", "XL", "M", "L", "M", "S", "L",
               "M", "XXL", "L", "M", "S", "M", "L"]
print(f"Most ordered shirt size: {mode(shirt_sizes)}")
print("→ This tells Zomato/Flipkart what to stock most of.")
````

## 4. Frequency Distribution — Seeing All Three at Once

A frequency distribution shows how often each value appears. It makes mean, median, and mode visually obvious.

````python
def frequency_distribution(data, bar_width=30):
    """
    Print an ASCII frequency distribution bar chart.
    Works for numbers and strings.
    """
    freq = {}
    for item in data:
        freq[item] = freq.get(item, 0) + 1

    max_count = max(freq.values())
    total     = len(data)

    print(f"  {'Value':<10} {'Count':>6} {'%':>6}  {'Bar'}")
    print("  " + "─" * (bar_width + 28))
    for value, count in sorted(freq.items(), key=lambda x: -x[1]):
        bar = "█" * int(count / max_count * bar_width)
        print(f"  {str(value):<10} {count:>6} {count/total*100:>5.1f}%  {bar}")


# Grade distribution for a class of 40
import random
random.seed(42)
raw_scores = [random.randint(40, 100) for _ in range(40)]
grade_data = []
for s in raw_scores:
    if s >= 80:   grade_data.append("A")
    elif s >= 60: grade_data.append("B")
    elif s >= 40: grade_data.append("C")
    else:         grade_data.append("F")

print("Grade distribution (40 students):")
frequency_distribution(grade_data)
print(f"\n  Mode (most common grade): {mode(grade_data)}")
````

````python
# Numeric frequency distribution — exam scores rounded to nearest 10
score_bins = [round(s / 10) * 10 for s in raw_scores]  # round to nearest 10

print("Score distribution (binned to nearest 10):")
frequency_distribution(score_bins)

print(f"\n  Mean   : {mean(raw_scores):.1f}")
print(f"  Median : {median(raw_scores)}")
print(f"  Mode   : {mode(score_bins)}  (binned)")
````

## 5. When to Use Which?

| Situation | Best measure | Why |
|---|---|---|
| Exam scores (fair test, no extremes) | **Mean** | Symmetric data, no outliers |
| Salaries / house prices | **Median** | Skewed right — a few very high values |
| Most popular product / city / grade | **Mode** | Categorical or frequency-based question |
| Reaction time / age distribution | **Median** | Often skewed |
| Daily temperature over a month | **Mean** | Reasonably symmetric |
| Survey rating (1–5 stars) | **Mode** | Discrete, customers cluster at extremes |

````python
# Skew detection: when mean ≠ median, data is skewed
# The gap tells you which direction

datasets = {
    "Symmetric (exam scores)"  : [55, 60, 65, 70, 72, 74, 75, 76, 78, 80, 82, 85, 90],
    "Right-skewed (salaries)"  : [30000, 32000, 35000, 38000, 40000, 42000, 45000, 300000],
    "Left-skewed (easy test)"  : [55, 88, 89, 90, 91, 92, 93, 94, 95, 96, 97, 98],
}

print(f"  {'Dataset':<30} {'Mean':>8} {'Median':>8} {'Gap':>8} {'Skew'}")
print("  " + "─" * 70)
for label, data in datasets.items():
    m   = mean(data)
    med = median(data)
    gap = m - med
    skew = "right →" if gap > 1 else "← left" if gap < -1 else "symmetric"
    print(f"  {label:<30} {m:>8.1f} {med:>8.1f} {gap:>8.1f} {skew}")

print()
print("Rule of thumb:")
print("  mean > median → right-skewed (pulled up by high outliers)")
print("  mean < median → left-skewed  (pulled down by low outliers)")
````

## 6. Applied — `describe()` Function

Pandas has a `df.describe()` method that prints summary statistics in one call. You'll use it constantly in Modules 5–8. Here's how it works internally.

````python
def describe(data, label="data"):
    """
    Print a summary of descriptive statistics for a numeric list.
    Mirrors what pandas df.describe() shows.
    """
    if not data:
        print(f"  {label}: (empty)")
        return {}

    n          = len(data)
    data_mean  = mean(data)
    data_med   = median(data)
    data_mode  = mode(data)
    data_min   = min(data)
    data_max   = max(data)
    data_range = data_max - data_min

    # Quartiles (preview of Lesson 03 — percentiles)
    sorted_d = sorted(data)
    q1 = sorted_d[n // 4]
    q3 = sorted_d[3 * n // 4]

    gap  = data_mean - data_med
    skew = "right-skewed" if gap > 1 else "left-skewed" if gap < -1 else "roughly symmetric"

    stats = {"count": n, "mean": round(data_mean, 2), "median": data_med,
             "mode": data_mode, "min": data_min, "max": data_max,
             "range": data_range, "q1": q1, "q3": q3, "skew": skew}

    print(f"  ── {label} ──")
    print(f"  count  : {n}")
    print(f"  mean   : {data_mean:.2f}")
    print(f"  median : {data_med}")
    print(f"  mode   : {data_mode}")
    print(f"  min    : {data_min}")
    print(f"  max    : {data_max}")
    print(f"  range  : {data_range}")
    print(f"  Q1     : {q1}")
    print(f"  Q3     : {q3}")
    print(f"  skew   : {skew}")
    return stats


# Student marks across three subjects
maths   = [72, 85, 68, 91, 76, 88, 55, 79, 83, 67, 90, 74, 62, 88, 95]
science = [65, 78, 82, 70, 88, 61, 74, 90, 55, 77, 83, 69, 94, 72, 80]
english = [80, 72, 88, 65, 91, 76, 82, 70, 85, 78, 68, 90, 74, 83, 77]

for subject, marks in [("Maths", maths), ("Science", science), ("English", english)]:
    describe(marks, label=subject)
    print()
````

## ⚠️ Common Mistakes

````python
# ── Mistake 1: Median on unsorted data ────────────────────────────────────────

data = [91, 55, 83, 67, 72]   # unsorted

# ❌ Taking the middle index without sorting first
wrong_median = data[len(data) // 2]
print(f"Wrong median (no sort): {wrong_median}")

# ✅ Always sort first
print(f"Correct median        : {median(data)}")
print(f"Sorted                : {sorted(data)}  → middle is {sorted(data)[2]}")
````

````python
# ── Mistake 2: Sorting the original list in-place ─────────────────────────────

scores = [91, 55, 83, 67, 72]
print(f"Before: {scores}")

# ❌ sort() modifies the original — other code using scores will get wrong data
# scores.sort()   # in-place — BAD in a function

# ✅ Use sorted() which returns a new list
sorted_scores = sorted(scores)
print(f"After sorted(): original={scores}, new={sorted_scores}")
````

````python
# ── Mistake 3: Assuming mode returns a single value ───────────────────────────

bimodal = [1, 2, 2, 3, 3, 4]

result = mode(bimodal)
print(f"mode([1, 2, 2, 3, 3, 4]) = {result}  ← a list, not a single value")

# ❌ Treating mode as always a single number
# my_mode = mode(bimodal)  +  10     # TypeError: can only concatenate list to list

# ✅ Check if it's a single mode before using it as a number
modes = mode(bimodal)
if len(modes) == 1:
    print(f"Single mode: {modes[0]}")
else:
    print(f"Multiple modes: {modes}")
````

````python
# ── Mistake 4: Using mean for categorical data ────────────────────────────────

grades_numeric = {"A": 4, "B": 3, "C": 2, "D": 1, "F": 0}
student_grades = ["A", "B", "B", "A", "C", "B", "A", "F", "B", "A"]

# Converting to numbers just to find average is misleading:
nums = [grades_numeric[g] for g in student_grades]
avg  = mean(nums)
print(f"Mean of encoded grades: {avg:.2f}")
print(f"  → What does 2.90 mean? It's between B(3) and C(2) — hard to interpret.")
print()
print(f"Mode of grades: {mode(student_grades)}")
print(f"  → 'A' is the most common grade — immediately useful.")
````

## 📌 Key Takeaways

- **Mean is pulled by outliers; median is not.** When your data has extreme values (salaries, house prices, reaction times, income), always report the median as the "typical" value. The mean is appropriate only when the distribution is reasonably symmetric and free of outliers — like exam scores on a fair test.

- **Mode is the only measure that works for categorical data.** You cannot average department names or shirt sizes. Mode tells you the most common category, which is often the most actionable insight (what to stock, what grade most students get, what city customers live in).

- **The gap between mean and median reveals skew.** If `mean > median`, the distribution is right-skewed — a few high values are pulling the average up. If `mean < median`, it's left-skewed. A data scientist checks this gap before applying any model, because many algorithms assume the data is roughly symmetric.
