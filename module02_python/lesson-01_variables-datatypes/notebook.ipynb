## 1. What Is a Variable?

A variable is a name that points to a value stored in memory.

```
  student_name = "Priya"
      │              │
    name          value
  (the label)    (what's in the box)
```

The `=` sign is the **assignment operator** — it puts the value on the right into the box with the label on the left. It does NOT mean "equals" in the mathematical sense.

````python
# Creating your first variables
student_name  = "Priya"
student_age   = 20
student_score = 88.5
is_enrolled   = True

print(student_name)
print(student_age)
print(student_score)
print(is_enrolled)
````

````python
# Variables can be updated — the label moves to a new box
score = 75
print("Before:", score)

score = 82          # score now points to 82, not 75
print("After: ", score)

score = score + 5   # read current value, add 5, store back
print("Bonus: ", score)
````

````python
# Multiple assignment — Python lets you do this in one line
x, y, z = 10, 20, 30
print(x, y, z)

# Assign the same value to multiple variables
a = b = c = 0
print(a, b, c)
````

## 2. The Five Core Data Types

Python has five types you will use in almost every program:

| Type | Keyword | Example | Use in data science |
|---|---|---|---|
| Integer | `int` | `42`, `-7`, `0` | Counts, indices, discrete values |
| Float | `float` | `3.14`, `-0.5`, `88.0` | Measurements, prices, probabilities |
| String | `str` | `"Priya"`, `"Mumbai"` | Names, categories, text data |
| Boolean | `bool` | `True`, `False` | Flags, conditions, binary features |
| None | `NoneType` | `None` | Missing values, uninitialised state |

Let's look at each one in detail.

### 2.1 — `int` (Integer)

Whole numbers. No decimal point. Can be positive, negative, or zero. No size limit in Python.

````python
num_students  = 42
goals_scored  = 3
temperature_c = -5          # negative int
population    = 1_428_628_000  # underscores for readability (Python 3.6+)

print(type(num_students))   # <class 'int'>
print(type(population))

# Integer arithmetic always produces exact results
print(10 + 3)   # 13
print(10 - 3)   # 7
print(10 * 3)   # 30
print(10 // 3)  # 3  ← integer division (floor division)
print(10 % 3)   # 1  ← remainder (modulo)
print(10 ** 3)  # 1000 ← exponentiation
````

**Data science context:** `int` appears as row counts, class labels (0/1), loop counters, and any discrete measurements. In pandas DataFrames, integer columns have dtype `int64`.

### 2.2 — `float` (Floating-point number)

Numbers with a decimal point. Represent continuous measurements.

````python
height_cm     = 162.5
passing_rate  = 0.87        # 87%
price         = 1_299.99
temperature   = -3.5
scientific    = 1.5e6       # 1,500,000 (scientific notation)

print(type(height_cm))      # <class 'float'>

# Regular division always produces a float
print(10 / 3)   # 3.3333333333333335  ← always float
print(9 / 3)    # 3.0                 ← even when result is whole, still float

# Mixing int and float gives float
result = 5 + 2.0
print(result, type(result))  # 7.0  <class 'float'>
````

````python
# ⚠ The floating-point trap — every beginner hits this
print(0.1 + 0.2)          # Expected: 0.3
print(0.1 + 0.2 == 0.3)   # Expected: True — WRONG!

# Why: computers store numbers in binary; 0.1 cannot be represented exactly
# Fix: use round() for display, or compare with a tolerance
print(round(0.1 + 0.2, 2))             # 0.3
print(abs(0.1 + 0.2 - 0.3) < 1e-9)    # True — safe comparison
````

### 2.3 — `str` (String)

Text data. Enclosed in single quotes `'...'`, double quotes `"..."`, or triple quotes `"""..."""` for multi-line.

````python
name     = "Priya Sharma"
city     = 'Mumbai'
category = "Electronics"
empty    = ""            # empty string — valid

print(type(name))        # <class 'str'>
print(len(name))         # 12 — number of characters

# Strings can be joined (concatenated) with +
greeting = "Hello, " + name + "!"
print(greeting)

# f-strings — the cleanest way to embed variables in text (Python 3.6+)
age   = 20
score = 88.5
report = f"{name} is {age} years old and scored {score:.1f} marks."
print(report)
````

````python
# Useful string methods — you'll use these constantly in data cleaning
raw = "  Mumbai  "           # common in real datasets: leading/trailing spaces
print(repr(raw.strip()))     # "Mumbai"    — removes whitespace

text = "priya sharma"
print(text.upper())          # "PRIYA SHARMA"
print(text.title())          # "Priya Sharma"  ← most useful for name columns
print(text.capitalize())     # "Priya sharma"

sentence = "Data Science is Fun"
print(sentence.lower())                  # "data science is fun"
print(sentence.replace("Fun", "Great")) # "Data Science is Great"
print(sentence.split(" "))              # ['Data', 'Science', 'is', 'Fun']
print("Science" in sentence)             # True — membership test
````

````python
# Strings are indexed — each character has a position (starting at 0)
city = "Mumbai"
#        M  u  m  b  a  i
#        0  1  2  3  4  5   ← forward index
#       -6 -5 -4 -3 -2 -1   ← backward index

print(city[0])    # M   — first character
print(city[-1])   # i   — last character
print(city[1:4])  # umb — characters at index 1, 2, 3 (4 is excluded)
print(city[:3])   # Mum — from start to index 3
print(city[3:])   # bai — from index 3 to end
print(city[::-1]) # iabmuM — reversed string
````

### 2.4 — `bool` (Boolean)

Only two possible values: `True` or `False`. The foundation of all decision-making in Python.

**Important:** `True` and `False` are capitalised — Python is case-sensitive.

````python
is_enrolled  = True
has_passed   = False
is_premium   = True

print(type(is_enrolled))   # <class 'bool'>

# Comparison operators produce booleans
score = 75
print(score > 60)          # True
print(score == 100)        # False  ← == checks equality (not =)
print(score != 100)        # True   ← != means "not equal"
print(score >= 75)         # True
print(score < 40)          # False
````

````python
# Booleans combine with logical operators: and, or, not
score        = 75
attendance   = 80   # percent

passed_exam  = score >= 40
good_attend  = attendance >= 75

# Both conditions must be True
eligible_for_grade = passed_exam and good_attend
print(f"Passed exam      : {passed_exam}")
print(f"Good attendance  : {good_attend}")
print(f"Eligible         : {eligible_for_grade}")

# At least one condition must be True
needs_support = score < 40 or attendance < 75
print(f"Needs support    : {needs_support}")

# Flip a boolean
print(f"Not enrolled     : {not is_enrolled}")
````

````python
# Booleans are secretly integers: True == 1, False == 0
# This is used constantly in data science to count conditions
scores = [85, 42, 91, 38, 77, 60, 55, 88]

# Count how many students passed (score >= 40)
passed_flags = [score >= 40 for score in scores]   # list of True/False
print("Pass flags:", passed_flags)
print("Passed    :", sum(passed_flags))             # sum counts Trues
print("Pass rate :", f"{sum(passed_flags)/len(scores)*100:.0f}%")
````

### 2.5 — `None` (NoneType)

`None` represents the **absence of a value**. It is not zero, not an empty string, not False. It means "nothing is here".

This is one of the most important types to understand for data science — missing values in datasets are almost always represented as `None` (or `NaN` in Pandas, which you'll learn in Module 5).

````python
# None represents a missing or unknown value
email           = None    # user hasn't provided email yet
secondary_score = None    # student didn't take the second exam
last_login      = None    # account never logged in

print(email)              # None
print(type(email))        # <class 'NoneType'>

# Always use 'is None' or 'is not None' to check — not ==
print(email is None)      # True  ← correct way
print(email == None)      # True  ← works but bad practice
print(email is not None)  # False
````

````python
# None vs zero vs empty string — these are DIFFERENT things
a = None    # unknown — data was never recorded
b = 0       # known — the student scored zero
c = ""      # known — the name field was left blank intentionally

print(f"None is falsy  : {bool(a)}")   # False
print(f"0 is falsy     : {bool(b)}")   # False
print(f"'' is falsy    : {bool(c)}")   # False
print()
print(f"None == 0      : {a == b}")    # False — they are not equal
print(f"None == ''     : {a == c}")    # False
print(f"0 == ''        : {b == c}")    # False
print()
# The distinction matters: a student who scored 0 and a student
# with no recorded score are two completely different situations
````

````python
# Handling None in practice — safe operations
student_scores = [88, None, 72, None, 91, 60, None, 45]

# Count missing values
missing = sum(1 for s in student_scores if s is None)
print(f"Missing scores : {missing} out of {len(student_scores)}")

# Compute average ignoring None
valid_scores = [s for s in student_scores if s is not None]
average      = sum(valid_scores) / len(valid_scores)
print(f"Valid scores   : {valid_scores}")
print(f"Average (excl. missing): {average:.1f}")
````

## 3. Checking Types with `type()`

````python
# type() tells you what Python type a value is
values = [
    42,
    3.14,
    "hello",
    True,
    None,
    "42",        # ← string that looks like a number
    "3.14",
    "True",
]

print(f"{'Value':<12}  {'type()'}")
print("─" * 30)
for v in values:
    print(f"{str(v):<12}  {type(v).__name__}")
````

````python
# isinstance() — check if a value is of a specific type (preferred in real code)
score = 88

print(isinstance(score, int))          # True
print(isinstance(score, float))        # False
print(isinstance(score, (int, float))) # True ← check multiple types at once
````

## 4. Type Conversion (Casting)

Real-world data is messy. A CSV file stores everything as text — even numbers. You'll constantly need to convert types.

Python provides built-in conversion functions: `int()`, `float()`, `str()`, `bool()`

````python
# ── str → int / float ─────────────────────────────────────────────────────────
# The most common conversion in data science: CSV values come in as strings

age_str   = "20"          # came from CSV
price_str = "1299.99"     # came from CSV

age   = int(age_str)
price = float(price_str)

print(f"age_str  : {age_str!r:10}  type: {type(age_str).__name__}")
print(f"age      : {age!r:10}  type: {type(age).__name__}")
print(f"price_str: {price_str!r:10}  type: {type(price_str).__name__}")
print(f"price    : {price!r:10}  type: {type(price).__name__}")

# Now arithmetic works
discounted = price * 0.9
print(f"\n10% discount: ₹{discounted:.2f}")
````

````python
# ── int / float → str ─────────────────────────────────────────────────────────
# Needed when building output messages or writing to files

score = 88
grade = "A"

# ❌ This fails — can't concatenate str and int directly
try:
    message = "Score: " + score
except TypeError as e:
    print(f"Error: {e}")

# ✅ Fix 1: convert to str first
message = "Score: " + str(score)
print(message)

# ✅ Fix 2 (better): use an f-string — no conversion needed
message = f"Score: {score}, Grade: {grade}"
print(message)
````

````python
# ── int ↔ float ───────────────────────────────────────────────────────────────
x = 7
y = float(x)     # 7.0 — no data lost

a = 3.9
b = int(a)       # 3   — decimal part TRUNCATED (not rounded!)
c = round(a)     # 4   — use round() if you want rounding

print(f"float({x}) → {y}")
print(f"int({a})   → {b}  ← truncated, not rounded!")
print(f"round({a}) → {c}")
````

````python
# ── bool() conversion — what counts as True and False ─────────────────────────
# This is called "truthiness" — very useful in data cleaning

test_values = [
    (1,       "int 1"),
    (0,       "int 0"),
    (-5,      "int -5"),
    (3.14,    "float 3.14"),
    (0.0,     "float 0.0"),
    ("hello", "non-empty str"),
    ("",      "empty str"),
    (None,    "None"),
]

print(f"{'Value':<12}  {'Description':<16}  {'bool()'}")
print("─" * 42)
for val, desc in test_values:
    print(f"{str(val):<12}  {desc:<16}  {bool(val)}")

print()
print("Rule: 0, 0.0, '', None → False. Everything else → True.")
````

````python
# ── Conversion failures — know what breaks before it breaks ───────────────────
conversions = [
    ("int('42')",     lambda: int("42"),     True),
    ("int('42.5')",   lambda: int("42.5"),   False),  # can't skip float step
    ("int('hello')",  lambda: int("hello"),  False),
    ("float('3.14')", lambda: float("3.14"), True),
    ("float('abc')",  lambda: float("abc"),  False),
    ("int(True)",     lambda: int(True),     True),
    ("int(False)",    lambda: int(False),    True),
]

print(f"{'Expression':<22}  {'Result'}")
print("─" * 45)
for expr, fn, should_work in conversions:
    try:
        result = fn()
        print(f"{expr:<22}  → {result}")
    except (ValueError, TypeError) as e:
        print(f"{expr:<22}  ✗ Error: {e}")
````

**Key rule for `int()` from string:** The string must contain only digits. `"42.5"` fails because it's not an integer. Convert via float first: `int(float("42.5"))` → `42`.

## 5. Dynamic Typing — Python's Superpower (and Gotcha)

In many languages (Java, C++), you must declare the type of a variable and it can never change:
```java
int score = 85;   // Java — locked as int forever
score = "A";      // ERROR
```

Python is **dynamically typed** — variables can hold any type and can change type at any time:

````python
# Python changes the type of x with each assignment
x = 10
print(f"x = {x!r:<12}  type: {type(x).__name__}")

x = 3.14
print(f"x = {x!r:<12}  type: {type(x).__name__}")

x = "hello"
print(f"x = {x!r:<12}  type: {type(x).__name__}")

x = True
print(f"x = {x!r:<12}  type: {type(x).__name__}")

x = None
print(f"x = {x!r:<12}  type: {type(x).__name__}")
````

**Why this matters in data science:**  
Dynamic typing makes Python fast to write but means *you* are responsible for knowing what type is in a variable at any given moment. Always use meaningful names and `type()` checks when something behaves unexpectedly.

## 7. Common Mistakes — and How to Fix Them

````python
# ── Mistake 1: Using = instead of == for comparison ──────────────────────────
score = 75

# ❌ This ASSIGNS 40 to score — doesn't check anything
# if score = 40:   ← SyntaxError — Python catches this one

# ✅ Use == to compare
if score == 75:
    print("Score is exactly 75")
````

````python
# ── Mistake 2: Concatenating str and int ─────────────────────────────────────
age = 20

# ❌ TypeError — can't add str and int
try:
    print("Age is " + age)
except TypeError as e:
    print(f"Error: {e}")

# ✅ Option A: str()
print("Age is " + str(age))

# ✅ Option B: f-string (cleaner)
print(f"Age is {age}")
````

````python
# ── Mistake 3: Operating on a string that looks like a number ─────────────────
# Extremely common when data is loaded from a CSV or user input

score_from_csv = "88"    # loaded as string

# ❌ Produces "8810" not 98 — string repetition, not arithmetic!
wrong_result = score_from_csv + "10"
print(f"String + String: {wrong_result!r}  ← wrong!")

# ✅ Convert first, then calculate
score = int(score_from_csv)
right_result = score + 10
print(f"Int + Int      : {right_result}    ← correct")
````

````python
# ── Mistake 4: Comparing with None using == ──────────────────────────────────
value = None

# ❌ Works but is bad practice — could break with custom objects
print(value == None)    # True — but avoid

# ✅ Always use 'is' to check for None
print(value is None)    # True — correct
print(value is not None) # False
````

## 8. Putting It All Together — A Student Info Card

````python
# A realistic mini-program that uses all five types

# ── Input data (as it might arrive from a CSV or web form) ────────────────────
raw_name        = "  priya sharma  "   # str — messy, needs cleaning
raw_age         = "20"                 # str — needs conversion
raw_score       = "88.5"               # str — needs conversion
raw_enrolled    = "True"              # str — needs conversion
raw_email       = None                 # NoneType — missing
PASSING_SCORE   = 40                   # constant int

# ── Clean and convert ─────────────────────────────────────────────────────────
name            = raw_name.strip().title()           # str   — cleaned
age             = int(raw_age)                       # int
score           = float(raw_score)                   # float
is_enrolled     = raw_enrolled == "True"             # bool  — string → bool
email           = raw_email                          # NoneType — stays as is

# ── Derived values ────────────────────────────────────────────────────────────
has_passed      = score >= PASSING_SCORE             # bool
grade           = "A" if score >= 80 else "B" if score >= 60 else "C" if score >= 40 else "F"
email_display   = email if email is not None else "Not provided"

# ── Output ───────────────────────────────────────────────────────────────────
print("=" * 40)
print("       STUDENT INFORMATION CARD")
print("=" * 40)
print(f"  Name        : {name}")
print(f"  Age         : {age} years")
print(f"  Score       : {score:.1f} / 100")
print(f"  Grade       : {grade}")
print(f"  Passed      : {'Yes ✓' if has_passed else 'No ✗'}")
print(f"  Enrolled    : {'Yes' if is_enrolled else 'No'}")
print(f"  Email       : {email_display}")
print("=" * 40)
print()
print("Data types used in this card:")
fields = [("name", name), ("age", age), ("score", score),
          ("has_passed", has_passed), ("email", email)]
for field, val in fields:
    print(f"  {field:<12} → {type(val).__name__:<10}  value: {val!r}")
````

## ⚠️ Common Misconceptions

### Misconception 1: `"10" + 10` should work
**No.** Python does not silently convert types like JavaScript does. `"10" + 10` raises a `TypeError`. You must explicitly convert: `int("10") + 10` = `20`. This is a feature — it prevents silent bugs.

### Misconception 2: `int(3.9)` rounds to 4
**No.** `int()` truncates — it drops the decimal part regardless. `int(3.9)` = `3`, `int(-3.9)` = `-3`. Use `round()` when you need rounding.

### Misconception 3: Variables store values
**Technically no.** Variables store *references* (pointers) to objects in memory. The value `42` exists in memory; your variable is a label that points to it. When you reassign, the label moves — the old object eventually gets cleaned up. This matters when you'll learn about mutable vs immutable types in the Lists lesson.

## 📌 Key Takeaways

- **Python has five core types: `int`, `float`, `str`, `bool`, `NoneType`.** Each has different valid operations. Mix them up and you'll get `TypeError` — or worse, a silent wrong answer.

- **`type()` tells you what Python type a value is — not its statistical type.** A pin code stored as `int` is still nominal. Always combine Python's type system with domain knowledge.

- **`None` means absence of value — not zero, not empty string.** Always check with `is None`, not `== None`. Missing values are everywhere in real datasets, and handling them correctly is a core data cleaning skill.
