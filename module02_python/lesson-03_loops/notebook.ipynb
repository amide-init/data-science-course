## 1. The `for` Loop

A `for` loop repeats a block of code **once for each item** in a sequence.

```python
for variable in sequence:
    # this block runs once per item
    do_something(variable)
```

The variable (`variable`) automatically takes the value of each item in turn. You choose the name — make it descriptive.

````python
# Iterating over a list
scores = [85, 42, 91, 38, 77, 60]

for score in scores:
    print(score)
````

````python
# Doing something useful in each iteration
scores = [85, 42, 91, 38, 77, 60]
total  = 0

for score in scores:
    total += score           # accumulate running total

average = total / len(scores)
print(f"Total   : {total}")
print(f"Average : {average:.1f}")
````

````python
# Iterating over a string — every character is an item
word = "Mumbai"
for char in word:
    print(char, end=" ")   # end=" " keeps output on one line
print()                    # newline after the loop

# Count vowels — a common text-processing task
sentence = "Data Science is the future"
vowels   = 0
for char in sentence.lower():
    if char in "aeiou":
        vowels += 1
print(f"Vowels in '{sentence}': {vowels}")
````

````python
# Iterating over a dictionary
student = {"name": "Priya", "age": 20, "score": 88.5, "city": "Mumbai"}

# .items() gives you (key, value) pairs
print("Student record:")
for key, value in student.items():
    print(f"  {key:<8}: {value}")
````

## 2. `range()` — Controlling the Number of Iterations

`range()` generates a sequence of numbers without storing them all in memory. It's the standard way to loop a specific number of times.

```python
range(stop)              # 0, 1, 2, ..., stop-1
range(start, stop)       # start, start+1, ..., stop-1
range(start, stop, step) # start, start+step, ..., up to stop-1
```

**Remember:** `range()` never includes the `stop` value — it stops just before it.

````python
# range(stop) — starts at 0
print("range(5):", list(range(5)))

# range(start, stop)
print("range(1, 6):", list(range(1, 6)))

# range(start, stop, step)
print("range(0, 20, 5):", list(range(0, 20, 5)))   # every 5th
print("range(10, 0, -1):", list(range(10, 0, -1))) # countdown
print("range(0, 10, 2):", list(range(0, 10, 2)))   # even numbers only
````

````python
# Common use: repeat something N times
print("Weekly report" + " ─" * 15)
for week in range(1, 5):                     # weeks 1 through 4
    print(f"  Week {week}: processing...")

print()
# Create a simple multiplication table
n = 5
print(f"Multiplication table for {n}:")
for i in range(1, 11):
    print(f"  {n} × {i:>2} = {n * i:>3}")
````

````python
# Practical: use range to access list items by index
students = ["Priya", "Rohan", "Meera", "Karan", "Ananya"]

# ❌ Avoid: range(len(...)) is clunky — works but isn't Pythonic
for i in range(len(students)):
    print(f"  {i}: {students[i]}")

print()
# ✅ Better: use enumerate() — shown in the next section
````

## 3. `enumerate()` — Loop with Index and Value

When you need both the position (index) and the item, use `enumerate()`. It's cleaner than `range(len(...))`.

````python
students = ["Priya", "Rohan", "Meera", "Karan", "Ananya"]

# enumerate() gives (index, value) pairs
for i, name in enumerate(students):
    print(f"  {i}: {name}")

print()
# Start numbering from 1 instead of 0
print("Rank list:")
for rank, name in enumerate(students, start=1):
    print(f"  {rank}. {name}")
````

````python
# Applied: find which rows have missing values
dataset = [
    {"name": "Priya",  "score": 88},
    {"name": "Rohan",  "score": None},   # missing
    {"name": "Meera",  "score": 72},
    {"name": None,     "score": 65},     # missing
    {"name": "Karan",  "score": 91},
]

print("Rows with missing values:")
for row_num, row in enumerate(dataset, start=1):
    missing_fields = [k for k, v in row.items() if v is None]
    if missing_fields:
        print(f"  Row {row_num}: missing {missing_fields}")
````

## 4. `zip()` — Loop Over Two Sequences in Parallel

`zip()` pairs up items from two (or more) sequences. Essential when your data is split across multiple lists.

````python
names  = ["Priya", "Rohan", "Meera", "Karan"]
scores = [88,      72,      95,      61     ]
cities = ["Mumbai", "Delhi", "Pune", "Chennai"]

# Pair up names and scores
for name, score in zip(names, scores):
    grade  = "Pass" if score >= 40 else "Fail"
    print(f"  {name:<8}  {score:>3}  {grade}")
````

````python
# zip() with three lists
for name, score, city in zip(names, scores, cities):
    print(f"  {name} ({city}): {score}")

print()
# zip() stops at the shortest list — be careful
a = [1, 2, 3, 4, 5]
b = ["a", "b", "c"]         # shorter
for x, y in zip(a, b):
    print(f"  {x} → {y}")   # only 3 pairs produced
````

````python
# Applied: combine split columns into records
# This happens constantly when loading CSV data column by column

student_ids = [1001, 1002, 1003, 1004]
names       = ["Priya", "Rohan", "Meera", "Karan"]
scores      = [88, 72, 95, 61]

records = []
for sid, name, score in zip(student_ids, names, scores):
    records.append({"id": sid, "name": name, "score": score})

print("Assembled records:")
for r in records:
    print(f"  {r}")
````

## 5. Nested Loops

A loop inside a loop. Each iteration of the outer loop runs the entire inner loop.

````python
# Nested loop: print a multiplication table
print("   ", end="")
for j in range(1, 6):
    print(f"{j:>4}", end="")
print()
print("   " + "─" * 20)

for i in range(1, 6):        # outer loop: rows
    print(f"{i} |", end="")
    for j in range(1, 6):    # inner loop: columns
        print(f"{i*j:>4}", end="")
    print()                  # newline after each row
````

````python
# Applied: cross-tabulation — count combinations
# How many students in each city × grade combination?

dataset = [
    {"city": "Mumbai", "grade": "A"},
    {"city": "Delhi",  "grade": "B"},
    {"city": "Mumbai", "grade": "A"},
    {"city": "Pune",   "grade": "C"},
    {"city": "Delhi",  "grade": "A"},
    {"city": "Mumbai", "grade": "B"},
    {"city": "Pune",   "grade": "A"},
    {"city": "Delhi",  "grade": "C"},
]

cities = ["Mumbai", "Delhi", "Pune"]
grades = ["A", "B", "C"]

# Header
print(f"{'City':<10}", end="")
for g in grades:
    print(f"  Grade {g}", end="")
print()
print("─" * 32)

# Count using nested loop
for city in cities:
    print(f"{city:<10}", end="")
    for grade in grades:
        count = sum(1 for r in dataset if r["city"] == city and r["grade"] == grade)
        print(f"  {count:>7}", end="")
    print()
````

## 6. The `while` Loop

A `while` loop repeats **as long as a condition is `True`**.

```python
while condition:
    # runs as long as condition is True
    do_something()
```

Use `while` when you don't know in advance how many iterations you need. Use `for` when iterating over a known sequence.

````python
# Basic while loop — countdown
count = 5
while count > 0:
    print(f"  {count}...")
    count -= 1       # must update count, otherwise infinite loop!
print("  Launch!")
````

````python
# While loop: keep searching until a condition is met
# Simulates reading records from a data stream until an error row is found

records = [
    {"id": 1, "value": 100},
    {"id": 2, "value": 250},
    {"id": 3, "value": -99},   # sentinel: -99 means end of valid data
    {"id": 4, "value": 180},
    {"id": 5, "value": 320},
]

i = 0
total = 0
while i < len(records) and records[i]["value"] != -99:
    total += records[i]["value"]
    i += 1

print(f"Processed {i} records before sentinel.")
print(f"Total value: {total}")
````

````python
# Applied: binary search — classic algorithm using while
# Find a value in a sorted list efficiently

def binary_search(sorted_list, target):
    low, high = 0, len(sorted_list) - 1
    steps     = 0

    while low <= high:
        mid = (low + high) // 2
        steps += 1

        if sorted_list[mid] == target:
            return mid, steps          # found!
        elif sorted_list[mid] < target:
            low = mid + 1              # target is in the right half
        else:
            high = mid - 1            # target is in the left half

    return -1, steps                  # not found

scores = sorted([42, 55, 61, 72, 78, 83, 88, 91, 95, 99])
print(f"Sorted scores: {scores}")
print()
for target in [83, 72, 50]:
    idx, steps = binary_search(scores, target)
    if idx != -1:
        print(f"  Found {target} at index {idx} in {steps} step(s)")
    else:
        print(f"  {target} not found (searched in {steps} step(s))")
````

## 7. Loop Control: `break` and `continue`

These keywords let you change the normal flow of a loop.

- `break` → **exit the loop immediately**
- `continue` → **skip the rest of this iteration, go to the next**

````python
# break: stop as soon as you find what you're looking for
students = [
    {"name": "Priya",  "score": 88},
    {"name": "Rohan",  "score": 72},
    {"name": "Meera",  "score": 95},
    {"name": "Karan",  "score": 61},
    {"name": "Ananya", "score": 91},
]

# Find the first student who scored > 90
top_scorer = None
for s in students:
    if s["score"] > 90:
        top_scorer = s
        break           # no need to keep looking

if top_scorer:
    print(f"First student scoring > 90: {top_scorer['name']} ({top_scorer['score']})")
````

````python
# continue: skip invalid / unwanted items and keep going
raw_scores = [88, None, -5, 72, None, 95, 110, 61]
# -5 and 110 are out of range; None is missing

valid_scores = []
skipped      = 0

for score in raw_scores:
    if score is None:          # skip missing
        skipped += 1
        continue
    if not (0 <= score <= 100): # skip out-of-range
        skipped += 1
        continue
    valid_scores.append(score) # only valid scores reach here

print(f"Raw    : {raw_scores}")
print(f"Valid  : {valid_scores}")
print(f"Skipped: {skipped} invalid / missing values")
print(f"Average: {sum(valid_scores)/len(valid_scores):.1f}")
````

````python
# break and continue together: process transactions, stop on error
transactions = [250, 1800, 500, -1, 3000, 750]   # -1 is a sentinel (error marker)

total   = 0
flagged = 0

SUSPICIOUS_THRESHOLD = 1000

for txn in transactions:
    if txn < 0:                   # sentinel — stop processing
        print(f"  Sentinel (-1) found. Stopping.")
        break
    if txn > SUSPICIOUS_THRESHOLD:
        print(f"  ⚠ Flagged: ₹{txn:,} (suspicious)")
        flagged += 1
        continue                  # still count it but move on
    total += txn

print(f"\nNormal total : ₹{total:,}")
print(f"Flagged txns : {flagged}")
````

## 8. List Comprehensions — Compact Loops

List comprehensions are a Pythonic shorthand for creating a new list by transforming or filtering an existing one.

```python
# Standard loop:
result = []
for item in sequence:
    result.append(expression)

# Comprehension — same thing, one line:
result = [expression  for item in sequence]

# With filter:
result = [expression  for item in sequence  if condition]
```

````python
scores = [88, 42, 72, 95, 61, 38, 80]

# Transform: add 5 bonus marks to every score
# Standard loop:
bonus_scores_loop = []
for s in scores:
    bonus_scores_loop.append(s + 5)

# Comprehension:
bonus_scores = [s + 5 for s in scores]

print(f"Original : {scores}")
print(f"Bonus    : {bonus_scores}")
print(f"Same?    : {bonus_scores_loop == bonus_scores}")
````

````python
# Filter: keep only passing scores (>= 40)
passing = [s for s in scores if s >= 40]
failing = [s for s in scores if s < 40]

print(f"All scores : {scores}")
print(f"Passing    : {passing}")
print(f"Failing    : {failing}")
````

````python
# Transform + filter at once
names = ["  priya  ", "ROHAN", "meera ", "  KARAN", "Ananya"]

# Clean: strip whitespace AND convert to title case, keep only names longer than 4 chars
cleaned = [name.strip().title() for name in names if len(name.strip()) > 4]
print(f"Raw     : {names}")
print(f"Cleaned : {cleaned}")
````

````python
# Common data science uses of comprehensions

dataset = [
    {"name": "Priya",  "score": 88, "city": "Mumbai"},
    {"name": "Rohan",  "score": 42, "city": "Delhi"},
    {"name": "Meera",  "score": 95, "city": "Mumbai"},
    {"name": "Karan",  "score": 61, "city": "Pune"},
    {"name": "Ananya", "score": 38, "city": "Delhi"},
]

# Extract one column
all_scores  = [row["score"]  for row in dataset]
all_names   = [row["name"]   for row in dataset]
mumbai_rows = [row for row in dataset if row["city"] == "Mumbai"]

# Add a derived column
with_grade = [
    {**row, "grade": "A" if row["score"] >= 80 else "B" if row["score"] >= 60 else "C" if row["score"] >= 40 else "F"}
    for row in dataset
]

print(f"All scores  : {all_scores}")
print(f"Mumbai rows : {[r['name'] for r in mumbai_rows]}")
print()
print("With grade column:")
for r in with_grade:
    print(f"  {r['name']:<8}  score: {r['score']}  grade: {r['grade']}")
````

## 9. Applied — Computing Statistics with Loops

Before you use NumPy or Pandas, let's see how all the basic statistics you'll use throughout the course are computed with pure Python loops. Understanding this makes the library functions feel less like magic.

````python
# Build a statistical summary function from scratch

def summarise(values, label="values"):
    """
    Compute key statistics for a list of numbers.
    Handles None (missing values) gracefully.
    """
    valid = [v for v in values if v is not None]   # drop missing

    if not valid:
        print(f"{label}: no valid data")
        return

    n        = len(valid)
    missing  = len(values) - n
    total    = sum(valid)
    mean     = total / n
    minimum  = min(valid)
    maximum  = max(valid)
    data_range = maximum - minimum

    # Median: sort and pick middle value
    sorted_vals = sorted(valid)
    mid = n // 2
    if n % 2 == 1:
        median = sorted_vals[mid]
    else:
        median = (sorted_vals[mid - 1] + sorted_vals[mid]) / 2

    # Variance and standard deviation
    variance = sum((v - mean) ** 2 for v in valid) / n
    std_dev  = variance ** 0.5

    print(f"─── {label} ──────────────────────")
    print(f"  Count     : {n}  (missing: {missing})")
    print(f"  Sum       : {total:.2f}")
    print(f"  Mean      : {mean:.2f}")
    print(f"  Median    : {median:.2f}")
    print(f"  Std Dev   : {std_dev:.2f}")
    print(f"  Min       : {minimum}")
    print(f"  Max       : {maximum}")
    print(f"  Range     : {data_range}")


exam_scores = [85, None, 72, 91, 38, 60, None, 77, 55, 88]
salaries    = [45000, 62000, 38000, 75000, 55000, 90000, 48000]

summarise(exam_scores, "Exam Scores")
print()
summarise(salaries, "Salaries (₹)")
````

## 10. Applied — Frequency Counter

Counting how many times each value appears is one of the most common operations in EDA. Here's how to build one from scratch.

````python
# Count category frequencies from a list

def frequency_table(data, label=""):
    """
    Counts occurrences of each unique value.
    Prints a sorted frequency table with percentage bar.
    """
    counts = {}
    for item in data:
        if item in counts:
            counts[item] += 1
        else:
            counts[item] = 1

    total = sum(counts.values())

    # Sort by frequency descending
    sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)

    title = label or "Frequency Table"
    print(f"─── {title} (n={total}) " + "─" * 20)
    print(f"  {'Category':<18}  {'Count':>5}  {'%':>6}  {'Bar'}")
    print(f"  {'─' * 18}  {'─' * 5}  {'─' * 6}  {'─' * 20}")
    for category, count in sorted_counts:
        pct = count / total * 100
        bar = "█" * int(pct / 5)    # each █ = 5%
        print(f"  {str(category):<18}  {count:>5}  {pct:>5.1f}%  {bar}")


# Student grade distribution
grades = ["A", "B", "A", "C", "B", "A", "F", "B", "C", "A",
          "B", "C", "A", "B", "D", "C", "A", "B", "A", "C"]
frequency_table(grades, "Grade Distribution")

print()

# Expense categories
expenses = [
    "Food", "Transport", "Food", "Shopping", "Food", "Entertainment",
    "Transport", "Food", "Health", "Food", "Shopping", "Transport",
    "Entertainment", "Food", "Transport", "Food", "Shopping", "Food"
]
frequency_table(expenses, "Expense Categories")
````

## 11. Applied — Outlier Detection with IQR

Outliers are data points that are unusually far from the rest. The IQR (Interquartile Range) method is the standard statistical approach — and you can build it with just loops.

````python
def find_outliers_iqr(values, label="values"):
    """
    Uses the IQR method to find outliers.
    Outlier = any value below Q1 - 1.5*IQR  or  above Q3 + 1.5*IQR
    """
    valid = sorted([v for v in values if v is not None])
    n     = len(valid)

    # Quartiles: split the sorted list into quarters
    q1 = valid[n // 4]           # 25th percentile (simplified)
    q3 = valid[(3 * n) // 4]     # 75th percentile (simplified)
    iqr = q3 - q1

    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr

    outliers = [v for v in valid if v < lower_fence or v > upper_fence]
    normal   = [v for v in valid if lower_fence <= v <= upper_fence]

    print(f"─── Outlier Detection: {label} ──────────────")
    print(f"  Q1 = {q1},  Q3 = {q3},  IQR = {iqr}")
    print(f"  Lower fence : {lower_fence}")
    print(f"  Upper fence : {upper_fence}")
    print(f"  Normal values  ({len(normal)}) : {normal}")
    print(f"  Outliers       ({len(outliers)}) : {outliers}")
    return outliers


salaries = [35000, 40000, 42000, 45000, 47000, 50000,
            52000, 55000, 60000, 65000, 70000, 150000, 5000]
find_outliers_iqr(salaries, "Monthly Salary (₹)")
````

## ⚠️ Common Mistakes

````python
# ── Mistake 1: Off-by-one with range() ───────────────────────────────────────
# range(n) goes 0 to n-1, NOT 0 to n

# ❌ Trying to print 1 through 5
print("Wrong (range(5)):")
for i in range(5):
    print(f"  {i}", end=" ")
print()   # prints 0 1 2 3 4

# ✅ Correct
print("Correct (range(1, 6)):")
for i in range(1, 6):
    print(f"  {i}", end=" ")
print()
````

````python
# ── Mistake 2: Modifying a list while iterating over it ──────────────────────
# Never add or remove items from a list inside a for loop over that same list

scores = [85, 42, 91, 38, 77, 60]

# ❌ Removing while iterating — skips items silently!
for score in scores:
    if score < 40:
        scores.remove(score)   # modifies the list being iterated!

print(f"After (wrong): {scores}")  # 38 may not be removed correctly

# ✅ Correct: iterate over a copy, build a new list
scores = [85, 42, 91, 38, 77, 60]
scores = [s for s in scores if s >= 40]  # comprehension is cleanest
print(f"After (correct): {scores}")
````

````python
# ── Mistake 3: Forgetting to update the while condition ──────────────────────
# This creates an infinite loop — Python will run forever

# ❌ INFINITE LOOP — do not uncomment!
# count = 0
# while count < 5:
#     print(count)    # count never changes → loops forever

# ✅ Always update the variable that controls the while condition
count = 0
while count < 5:
    print(count, end=" ")
    count += 1      # this line is essential
print()
````

````python
# ── Mistake 4: Accumulator not initialised before the loop ───────────────────

scores = [85, 72, 91, 61]

# ❌ Will raise NameError — total doesn't exist yet
try:
    for s in scores:
        total_wrong += s   # NameError: total_wrong not defined
except NameError as e:
    print(f"Error: {e}")

# ✅ Always initialise before the loop
total = 0
for s in scores:
    total += s
print(f"Total: {total}")
````

````python
# ── Mistake 5: Using range(len(list)) when enumerate() is cleaner ─────────────

students = ["Priya", "Rohan", "Meera"]

# ❌ Clunky — works but is not Pythonic
print("range(len(...)):")
for i in range(len(students)):
    print(f"  {i + 1}. {students[i]}")

# ✅ Clean — use enumerate
print("enumerate():")
for i, name in enumerate(students, start=1):
    print(f"  {i}. {name}")
````

## 📌 Key Takeaways

- **Use `for` when you know the sequence; use `while` when you know the stopping condition.** In data science, `for` loops dominate — you almost always iterate over known datasets. `while` appears in algorithms and streaming data scenarios.

- **`enumerate()` and `zip()` are your everyday tools, not extras.** Reaching for `range(len(...))` is a sign that `enumerate()` would be cleaner. Parallel iteration over two lists is a `zip()` — don't synchronise them manually.

- **List comprehensions are loops — not magic.** They run exactly the same code as the equivalent `for` loop; they're just more concise. Use them for transforms and filters. Don't nest them more than one level deep or they become harder to read than the loop they replaced.
