## 1. Arithmetic Operators

You already know `+`, `-`, `*`, `/`. Python has three more that are critical for data work.

````python
# All seven arithmetic operators
a, b = 17, 5

print(f"{a} + {b}  = {a + b}    addition")
print(f"{a} - {b}  = {a - b}    subtraction")
print(f"{a} * {b}  = {a * b}    multiplication")
print(f"{a} / {b}  = {a / b}    division       (always float)")
print(f"{a} // {b} = {a // b}    floor division (drop decimal, stays int)")
print(f"{a} % {b}  = {a % b}    modulo         (remainder after division)")
print(f"{a} ** {b} = {a ** b}  exponentiation")
````

````python
# Floor division and modulo — the underrated pair
# Together they let you split a number into quotient and remainder

total_minutes = 137
hours   = total_minutes // 60    # how many complete hours
minutes = total_minutes % 60     # leftover minutes
print(f"{total_minutes} minutes = {hours}h {minutes}m")

# Modulo: check if a number is even or odd
for n in [4, 7, 12, 9]:
    parity = "even" if n % 2 == 0 else "odd"
    print(f"  {n} is {parity}")

# Modulo: useful in data science for splitting datasets
# e.g., every 5th row goes to validation set
print()
for i in range(10):
    split = "validation" if i % 5 == 0 else "training"
    print(f"  Row {i}: {split}")
````

### 1.1 Operator Precedence (BODMAS in Python)

When multiple operators appear together, Python follows a strict order:

```
Highest priority
  **          exponentiation
  -x          unary minus
  * / // %    multiplication, division, floor div, modulo
  + -         addition, subtraction
Lowest priority
```

When in doubt — **use parentheses**. They cost nothing and prevent bugs.

````python
# Precedence surprises
print(2 + 3 * 4)        # 14, not 20 — multiplication first
print((2 + 3) * 4)      # 20 — parentheses override
print(2 ** 3 ** 2)      # 512, not 64 — ** is right-to-left: 2**(3**2) = 2**9
print((2 ** 3) ** 2)    # 64

# A data science calculation that goes wrong without parentheses
scores = [80, 92, 71, 88]

# ❌ Wrong: only divides last element by 4
wrong_avg = scores[0] + scores[1] + scores[2] + scores[3] / 4
print(f"\nWrong average : {wrong_avg}")

# ✅ Correct: sum everything first, then divide
correct_avg = (scores[0] + scores[1] + scores[2] + scores[3]) / 4
print(f"Correct average: {correct_avg}")
````

### 1.2 Augmented Assignment Operators

Shortcuts for updating a variable using its current value.

````python
# These two lines do exactly the same thing:
# score = score + 10
# score += 10            ← augmented assignment

total = 0
total += 450     # total = total + 450
total += 299     # total = total + 299
total += 1199    # total = total + 1199
print(f"Total: ₹{total}")

# All variants
x = 20
x += 5;  print(f"+=  → {x}")
x -= 3;  print(f"-=  → {x}")
x *= 2;  print(f"*=  → {x}")
x //= 3; print(f"//= → {x}")
x %= 4;  print(f"%=  → {x}")
x **= 3; print(f"**= → {x}")
````

## 2. Comparison Operators

These always return a `bool` (`True` or `False`). They are the backbone of all data filtering.

````python
score = 72

print(f"score == 72  → {score == 72}   equal to")
print(f"score != 72  → {score != 72}  not equal to")
print(f"score >  60  → {score > 60}   greater than")
print(f"score <  60  → {score < 60}  less than")
print(f"score >= 72  → {score >= 72}   greater than or equal")
print(f"score <= 70  → {score <= 70}  less than or equal")
````

````python
# Python's chained comparisons — a beautiful feature
# Other languages require: score >= 60 and score < 80
# Python lets you write it as a mathematical range:

score = 72

# Both are equivalent — Python style is cleaner
print(score >= 60 and score < 80)   # traditional
print(60 <= score < 80)             # Python chained — reads like maths

# Real use: checking if a value is in a valid range
age         = 25
salary      = 55000
credit_score = 710

print()
print(f"Age in working range (18–60) : {18 <= age <= 60}")
print(f"Salary in middle band        : {30000 <= salary <= 80000}")
print(f"Credit score is good (700+)  : {credit_score >= 700}")
````

````python
# Comparing strings — uses alphabetical (lexicographic) order
print("apple" < "banana")     # True  — 'a' comes before 'b'
print("Mumbai" == "mumbai")   # False — case sensitive!
print("Mumbai" == "Mumbai")   # True

# Always normalise string case before comparing in data cleaning
city_input = "MUMBAI"
print(city_input.lower() == "mumbai")   # True — safe comparison
````

## 3. Logical Operators

`and`, `or`, `not` combine boolean expressions. Essential for multi-condition filtering.

````python
# Truth tables — what each operator produces
combos = [(True, True), (True, False), (False, True), (False, False)]

print(f"{'A':<7} {'B':<7} {'A and B':<10} {'A or B':<10} {'not A'}")
print("─" * 45)
for a, b in combos:
    print(f"{str(a):<7} {str(b):<7} {str(a and b):<10} {str(a or b):<10} {str(not a)}")
````

````python
# Applied: loan eligibility — multiple conditions must ALL be true
age          = 32
income       = 65000
credit_score = 720
employment   = "salaried"   # or "self-employed"
existing_loans = 1

age_ok      = 21 <= age <= 60
income_ok   = income >= 25000
credit_ok   = credit_score >= 650
employ_ok   = employment in ("salaried", "self-employed")
loans_ok    = existing_loans < 3

eligible = age_ok and income_ok and credit_ok and employ_ok and loans_ok

print("Loan eligibility check:")
print(f"  Age (21–60)            : {'✓' if age_ok    else '✗'}  {age}")
print(f"  Income (≥₹25,000)      : {'✓' if income_ok else '✗'}  ₹{income:,}")
print(f"  Credit score (≥650)    : {'✓' if credit_ok else '✗'}  {credit_score}")
print(f"  Employment type        : {'✓' if employ_ok else '✗'}  {employment}")
print(f"  Existing loans (< 3)   : {'✓' if loans_ok  else '✗'}  {existing_loans}")
print(f"  ─────────────────────────")
print(f"  ELIGIBLE               : {'✓ YES' if eligible else '✗ NO'}")
````

````python
# Short-circuit evaluation — Python stops as soon as result is known
# 'and': if first is False, second is never checked
# 'or' : if first is True,  second is never checked

# This is useful to avoid errors:
data = None

# ❌ Would crash: AttributeError — can't call .upper() on None
# if data.upper() == "MUMBAI":  ...

# ✅ Safe: if data is None, the second condition is never evaluated
if data is not None and data.upper() == "MUMBAI":
    print("City is Mumbai")
else:
    print("Data is missing or city is not Mumbai")

# 'or' short-circuit: use a default value when something might be None
user_city   = None
default_city = "Mumbai"
city = user_city or default_city    # if user_city is falsy, use default
print(f"City used: {city}")
````

## 4. Membership Operators: `in` and `not in`

Check whether a value exists within a collection. One of the most readable features of Python.

````python
# 'in' with strings — check if a substring exists
email = "priya@gmail.com"
print("@" in email)           # True  — valid email check (basic)
print(".com" in email)        # True
print("gmail" in email)       # True
print("yahoo" in email)       # False

# 'in' with lists and tuples — check membership
valid_categories = ["Food", "Transport", "Entertainment", "Shopping", "Health"]
user_category    = "Shopping"
unknown_category = "Gambling"

print()
print(f"'{user_category}' in valid list    : {user_category in valid_categories}")
print(f"'{unknown_category}' in valid list : {unknown_category in valid_categories}")
````

````python
# Applied: data validation — flag rows with invalid categories
VALID_GRADES      = {"A", "B", "C", "D", "F"}   # set is fastest for membership
VALID_DEPARTMENTS = {"CS", "EC", "ME", "CE", "IT"}

student_records = [
    {"name": "Priya",  "grade": "A",  "dept": "CS"},
    {"name": "Rohan",  "grade": "B",  "dept": "ME"},
    {"name": "Meera",  "grade": "X",  "dept": "CS"},   # invalid grade
    {"name": "Karan",  "grade": "C",  "dept": "BA"},   # invalid dept
    {"name": "Ananya", "grade": "Z",  "dept": "XY"},   # both invalid
]

print(f"{'Name':<10} {'Grade':<8} {'Dept':<6} {'Status'}")
print("─" * 45)
for s in student_records:
    grade_ok = s["grade"] in VALID_GRADES
    dept_ok  = s["dept"]  in VALID_DEPARTMENTS
    issues   = []
    if not grade_ok: issues.append(f"invalid grade '{s['grade']}'")
    if not dept_ok:  issues.append(f"invalid dept '{s['dept']}'")
    status = "✓ OK" if not issues else "✗ " + ", ".join(issues)
    print(f"{s['name']:<10} {s['grade']:<8} {s['dept']:<6} {status}")
````

## 5. Identity Operators: `is` and `is not`

`is` checks if two variables point to the **exact same object in memory** — not just equal values.

````python
# The main use case: checking for None
value = None
print(value is None)       # True  ← always use 'is' for None
print(value is not None)   # False

# Why 'is' instead of '=='?
# 'is' checks identity (same object), '==' checks equality (same value)
# For None, there is only ONE None object in Python — so 'is' is precise

# is vs == with booleans — same reason
flag = True
print(flag is True)     # True  — same object
print(flag == 1)        # True  — equal value (True == 1 numerically)
print(flag is 1)        # False — different objects (and a warning in Python 3.8+)
````

````python
# The golden rule:
# Use 'is' / 'is not' ONLY for:  None, True, False
# Use '==' / '!='  for everything else

score = 100

# ❌ Bad practice — don't use 'is' for numbers or strings
# (works for small integers due to caching but is not reliable)
print(score is 100)     # True for small ints but unreliable — avoid

# ✅ Always use == for value equality
print(score == 100)     # True — always reliable
````

## 6. Conditions — `if`, `elif`, `else`

Conditions let your code make decisions. They are the most fundamental control structure in programming.

### 6.1 Basic `if`

```python
if condition:
    # runs only when condition is True
    do_something()
```

**Indentation is NOT optional in Python.** Four spaces (or one tab) defines what's inside the `if`. Get it wrong and Python raises an `IndentationError`.

````python
score = 85

if score >= 90:
    print("Distinction — excellent work!")

# Nothing printed — condition was False, block is skipped entirely
print("This line always runs, regardless of the if block")
````

### 6.2 `if` / `else`

```python
if condition:
    # runs when True
else:
    # runs when False
```

````python
score = 38

if score >= 40:
    result = "Pass"
else:
    result = "Fail"

print(f"Score: {score}  →  {result}")

# Applied: categorise a transaction amount
amount = 2500

if amount > 1000:
    transaction_type = "large"
else:
    transaction_type = "small"

print(f"₹{amount} is a {transaction_type} transaction")
````

### 6.3 `if` / `elif` / `else` — Multiple Branches

```python
if condition_1:
    ...
elif condition_2:
    ...
elif condition_3:
    ...
else:
    ...  # catches everything that didn't match
```

Python checks conditions **top to bottom** and stops at the first `True`. Only one branch ever runs.

````python
# Grade calculator — a classic elif chain
def get_grade(score):
    if score >= 90:
        return "A+", "Distinction"
    elif score >= 80:
        return "A",  "First Class"
    elif score >= 70:
        return "B",  "Second Class"
    elif score >= 60:
        return "C",  "Pass Class"
    elif score >= 40:
        return "D",  "Pass"
    else:
        return "F",  "Fail"

test_scores = [95, 82, 74, 61, 45, 32]
print(f"{'Score':>6}  {'Grade':<5}  {'Category'}")
print("─" * 30)
for s in test_scores:
    grade, category = get_grade(s)
    print(f"{s:>6}  {grade:<5}  {category}")
````

````python
# The order of elif conditions matters — always check most specific first

# ❌ Wrong order — score 95 will match the FIRST true condition (>= 40)
def get_grade_wrong(score):
    if score >= 40:    # ← too broad — catches everything >= 40
        return "D"
    elif score >= 60:
        return "C"     # this code is unreachable!
    elif score >= 80:
        return "A"     # this code is unreachable!

print(f"Wrong  grader: score 95 → grade {get_grade_wrong(95)}")  # 'D' — wrong!
print(f"Correct grader: score 95 → grade {get_grade(95)[0]}")    # 'A+' — right

print()
print("Rule: put the MOST RESTRICTIVE condition first.")
print("More specific (>= 90) before less specific (>= 40).")
````

### 6.4 Nested Conditions

Conditions inside conditions. Use when a decision depends on another decision.

````python
# Scholarship eligibility — nested decision tree
def check_scholarship(score, attendance, income):
    if score >= 75:                         # first gate: academic
        if attendance >= 80:               # second gate: attendance
            if income < 300000:            # third gate: financial need
                return "Full scholarship"
            else:
                return "Merit scholarship (no financial need criteria)"
        else:
            return "Ineligible — attendance below 80%"
    else:
        return "Ineligible — score below 75"

students = [
    ("Priya",  88, 92, 250000),
    ("Rohan",  88, 92, 500000),  # good score, attendance but higher income
    ("Meera",  88, 70, 150000),  # good score but low attendance
    ("Karan",  60, 95, 200000),  # good attendance but low score
]

print(f"{'Name':<10} {'Score':>6} {'Attend':>7} {'Income':>10}  {'Result'}")
print("─" * 65)
for name, score, attend, income in students:
    result = check_scholarship(score, attend, income)
    print(f"{name:<10} {score:>6} {attend:>6}% {income:>10,}  {result}")
````

**When to avoid deep nesting:** More than 2–3 levels of nesting makes code hard to read. Often you can flatten it by returning early (`guard clauses`) or using `and` to combine conditions.

````python
# Flattening nested conditions with guard clauses + and
def check_scholarship_flat(score, attendance, income):
    # Reject early on any failing condition
    if score < 75:
        return "Ineligible — score below 75"
    if attendance < 80:
        return "Ineligible — attendance below 80%"
    # Only reach here if both above passed
    if income < 300000:
        return "Full scholarship"
    return "Merit scholarship (no financial need criteria)"

# Same results, much easier to read
print(check_scholarship_flat(88, 92, 250000))
print(check_scholarship_flat(60, 95, 200000))
````

### 6.5 Ternary Expression (One-line `if`)

```python
value = result_if_true  if  condition  else  result_if_false
```

Use only for simple, readable cases. Don't nest ternary expressions — it kills readability.

````python
score = 75

# These two blocks produce the same result:

# Multi-line version
if score >= 40:
    status = "Pass"
else:
    status = "Fail"
print(status)

# Ternary version — great for simple yes/no
status = "Pass" if score >= 40 else "Fail"
print(status)

# Common use: display formatting
is_premium = True
label = "⭐ Premium" if is_premium else "Standard"
print(label)

# Safe fallback for None values
email = None
display_email = email if email is not None else "Not provided"
print(display_email)
````

### 7.1 — Data Quality Checker

````python
# Before analysing any dataset, run each row through a quality check.
# This is a simplified version of what you'll build in Module 6.

def check_row_quality(row):
    """
    Returns a list of quality issues found in a student record.
    An empty list means the row is clean.
    """
    issues = []

    # Check: name must be a non-empty string
    if not isinstance(row.get("name"), str) or not row["name"].strip():
        issues.append("missing or empty name")

    # Check: age must be a realistic integer
    age = row.get("age")
    if age is None:
        issues.append("missing age")
    elif not isinstance(age, int) or not (15 <= age <= 35):
        issues.append(f"age out of range ({age})")

    # Check: score must be 0–100
    score = row.get("score")
    if score is None:
        issues.append("missing score")
    elif not isinstance(score, (int, float)) or not (0 <= score <= 100):
        issues.append(f"score out of range ({score})")

    # Check: email must contain '@'
    email = row.get("email")
    if email is not None and "@" not in str(email):
        issues.append(f"invalid email ({email})")

    return issues


# Test dataset with intentional errors
dataset = [
    {"name": "Priya",  "age": 20, "score": 88.5, "email": "priya@gmail.com"},
    {"name": "",       "age": 21, "score": 72.0, "email": "rohan@gmail.com"},   # empty name
    {"name": "Meera",  "age": 19, "score": 115,  "email": "meera@gmail.com"},   # score > 100
    {"name": "Karan",  "age": 99, "score": 60.0, "email": None},                # invalid age
    {"name": "Ananya", "age": 22, "score": 91.0, "email": "ananya_no_at_sign"}, # bad email
    {"name": None,     "age": None, "score": None, "email": None},              # all missing
]

print(f"{'Name':<10}  {'Status'}")
print("─" * 60)
clean, dirty = 0, 0
for row in dataset:
    name   = str(row.get("name") or "(no name)")
    issues = check_row_quality(row)
    if issues:
        print(f"{name:<10}  ✗ {'; '.join(issues)}")
        dirty += 1
    else:
        print(f"{name:<10}  ✓ OK")
        clean += 1

print(f"\nClean rows: {clean} / {len(dataset)}   Dirty rows: {dirty}")
````

### 7.2 — Discount Engine

````python
# E-commerce discount logic — exactly the kind of rule engine real platforms run

def calculate_discount(order_amount, is_premium, coupon_code, first_order):
    """
    Returns (discount_percent, reason) for a given order.
    Discounts don't stack — highest applicable discount wins.
    """
    valid_coupons = {
        "SAVE20": 20,
        "FLAT10": 10,
        "YTV15" : 15,
    }

    # Priority 1: first order gets the best deal
    if first_order:
        return 25, "First order discount"

    # Priority 2: valid coupon code
    if coupon_code in valid_coupons:
        pct = valid_coupons[coupon_code]
        return pct, f"Coupon '{coupon_code}'"

    # Priority 3: premium member perks
    if is_premium and order_amount >= 1000:
        return 15, "Premium member + large order"
    elif is_premium:
        return 10, "Premium member"

    # Priority 4: bulk order discount
    if order_amount >= 5000:
        return 12, "Bulk order (₹5000+)"
    elif order_amount >= 2000:
        return 8, "Large order (₹2000+)"

    # No discount
    return 0, "No discount applicable"


orders = [
    (800,  False, None,     True,  "New customer, small order"),
    (1500, True,  None,     False, "Premium, medium order"),
    (3000, False, "SAVE20", False, "Coupon user"),
    (6000, False, None,     False, "Bulk order, no membership"),
    (500,  False, "NOPE99", False, "Invalid coupon, small order"),
    (2500, True,  "YTV15",  False, "Premium + coupon"),
]

print(f"{'Amount':>8}  {'Discount':>9}  {'Savings':>8}  {'Reason'}")
print("─" * 65)
for amount, premium, coupon, first, label in orders:
    pct, reason = calculate_discount(amount, premium, coupon, first)
    savings     = amount * pct / 100
    final       = amount - savings
    print(f"₹{amount:>7,}  {pct:>7}%    ₹{savings:>6.0f}  {reason}")
````

### 7.3 — Transaction Fraud Rule Engine

````python
# A simplified rule-based fraud detector
# Real systems use ML — but rule engines are still the first layer of defence

def assess_fraud_risk(txn):
    """
    Evaluates a transaction against fraud rules.
    Returns (risk_level, triggered_rules).
    """
    rules_triggered = []
    risk_score      = 0

    # Rule 1: unusually large amount (3× the account average)
    if txn["amount"] > txn["account_avg"] * 3:
        rules_triggered.append("Amount 3× account average")
        risk_score += 40

    # Rule 2: transaction at unusual hour (1 AM – 5 AM)
    if 1 <= txn["hour"] <= 5:
        rules_triggered.append("Unusual hour (1–5 AM)")
        risk_score += 25

    # Rule 3: different country from user's home country
    if txn["merchant_country"] != txn["home_country"]:
        rules_triggered.append("Foreign merchant")
        risk_score += 20

    # Rule 4: high-risk merchant category
    HIGH_RISK_CATEGORIES = {"crypto", "gambling", "wire_transfer"}
    if txn["category"] in HIGH_RISK_CATEGORIES:
        rules_triggered.append(f"High-risk category ({txn['category']})")
        risk_score += 35

    # Rule 5: card used in two countries within 2 hours (impossible travel)
    if txn.get("impossible_travel", False):
        rules_triggered.append("Impossible travel detected")
        risk_score += 60

    # Risk classification
    if risk_score >= 70:
        level = "HIGH — BLOCK"
    elif risk_score >= 40:
        level = "MEDIUM — FLAG"
    elif risk_score > 0:
        level = "LOW — MONITOR"
    else:
        level = "CLEAR"

    return risk_score, level, rules_triggered


transactions = [
    {"id": "T001", "amount": 500,   "account_avg": 400, "hour": 14,
     "merchant_country": "IN", "home_country": "IN", "category": "food"},

    {"id": "T002", "amount": 45000, "account_avg": 3000, "hour": 3,
     "merchant_country": "US", "home_country": "IN", "category": "electronics"},

    {"id": "T003", "amount": 2000,  "account_avg": 4000, "hour": 21,
     "merchant_country": "IN", "home_country": "IN", "category": "gambling"},

    {"id": "T004", "amount": 800,   "account_avg": 600, "hour": 11,
     "merchant_country": "AE", "home_country": "IN", "category": "shopping",
     "impossible_travel": True},
]

for txn in transactions:
    score, level, rules = assess_fraud_risk(txn)
    print(f"Transaction {txn['id']} — ₹{txn['amount']:,}")
    print(f"  Risk score : {score}")
    print(f"  Decision   : {level}")
    if rules:
        print(f"  Rules hit  : {', '.join(rules)}")
    else:
        print(f"  No rules triggered")
    print()
````

## ⚠️ Common Misconceptions

````python
# Misconception 1: elif and else are mandatory
# Not true — if stands alone perfectly fine

score = 85
if score > 90:
    print("Distinction")   # just skipped if False — no else needed
print("Always prints")

# Misconception 2: multiple if chains vs elif chain — they're different
score = 85
print()
print("Multiple 'if' (WRONG for grading — prints multiple results):")
if score >= 80: print("  A")
if score >= 60: print("  B")   # ← also True! Both run.
if score >= 40: print("  C")   # ← also True!

print("elif chain (CORRECT — stops at first match):")
if   score >= 80: print("  A")
elif score >= 60: print("  B")  # ← skipped
elif score >= 40: print("  C")  # ← skipped
````

````python
# Misconception 3: 'and' always evaluates both sides — no
# Short-circuit means the second operand may be skipped

def risky_check():
    print("  risky_check was called")
    return True

print("False and risky_check():")
result = False and risky_check()   # risky_check never called
print(f"  result = {result}")

print()
print("True or risky_check():")
result = True or risky_check()    # risky_check never called
print(f"  result = {result}")
````

## 📌 Key Takeaways

- **Use parentheses to make precedence explicit.** The `*` before `+` rule is easy to forget in complex expressions. `(a + b) / n` is always clearer than relying on precedence.

- **`elif` chains stop at the first match — multiple `if` statements don't.** Use `elif` when exactly one branch should run (grading, categorisation). Use separate `if`s when multiple conditions can independently trigger actions.

- **`in` and `not in` are your first data filtering tools.** Checking category membership, validating allowed values, and searching for substrings all use `in`. When you reach Pandas in Module 5, every boolean filter column you build will use these same comparisons under the hood.
