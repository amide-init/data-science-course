> Picture a single column from a spreadsheet: a list of exam scores, each one lined up next to a student's name. That's precisely what a pandas **Series** is — a one-dimensional, labeled array. It's built directly on top of the NumPy array you've been working with, but with one addition that changes everything: every value gets a **label** (called the index), not just a position.

That label is the whole story of Pandas. Once every value has a name attached to it — not just "the 3rd item" — you can align data, merge datasets, and filter by meaning instead of position. Every column in a DataFrame (next lesson) is secretly a Series, so mastering it here pays off immediately.

## 1. Creating Your First Series

````python
import pandas as pd
import numpy as np

# A list of exam scores -- same data you'd use to build a NumPy array
scores = pd.Series([88, 92, 79, 95, 61])

print(scores)
print()
print(f"type: {type(scores)}")
````

Notice the two columns in the printed output: the **left** column (0, 1, 2, 3, 4) is the **index** -- automatically generated here since we didn't specify one. The **right** column is the actual data. A plain NumPy array only ever has that right column; the index is what pandas adds on top.

---
## 2. Giving the Index Meaning

The default 0, 1, 2... index is just a placeholder. The real power of a Series shows up when you give the index actual labels -- turning "item at position 2" into "Aisha's score."

````python
students = ["Aisha", "Ben", "Chen", "Diya", "Elan"]
scores = pd.Series([88, 92, 79, 95, 61], index=students)

print(scores)
print()
print(f"scores['Chen'] = {scores['Chen']}   <- accessed by LABEL, not position")
````

---
## 3. Creating a Series from a Dictionary

A Python dict already pairs labels (keys) with values -- which is exactly what a Series needs. Pandas converts dict keys directly into the index.

````python
grades_dict = {"Aisha": 88, "Ben": 92, "Chen": 79, "Diya": 95, "Elan": 61}
grades = pd.Series(grades_dict)

print(grades)
print()
print(f"grades.index  : {list(grades.index)}")
print(f"grades.values : {grades.values}   <- this part IS a plain NumPy array")
````

---
## 4. Creating a Series from a Scalar

A single value can be "broadcast" across a whole index -- useful for setting up a placeholder column before filling in real data.

````python
default_attendance = pd.Series(100, index=students)
print(default_attendance)
````

---
## 5. Accessing Values: `.loc` vs `.iloc`

This is the single most important distinction in this lesson:

- **`.loc[label]`** — access by index **label** (what you named it)
- **`.iloc[position]`** — access by integer **position** (0-based, like a list), regardless of what the labels are

They usually agree when the index is the default 0, 1, 2... but diverge completely once you use custom labels.

````python
grades = pd.Series([88, 92, 79, 95, 61], index=students)

print(f"grades.loc['Ben']  : {grades.loc['Ben']}    <- by LABEL 'Ben'")
print(f"grades.iloc[1]     : {grades.iloc[1]}    <- by POSITION 1 (also Ben here, same result)")
print()

# Slicing shows the real difference
print("grades.loc['Ben':'Diya']  (label slice, END IS INCLUDED):")
print(grades.loc["Ben":"Diya"])
print()
print("grades.iloc[1:4]  (position slice, END IS EXCLUDED, just like a Python list):")
print(grades.iloc[1:4])
````

---
## 6. Vectorized Arithmetic

Just like a NumPy array, a Series supports fast, whole-array math -- no manual loop required. Every value gets the operation applied at once, and the index labels come along for the ride.

````python
grades = pd.Series([88, 92, 79, 95, 61], index=students)

curved = grades + 5     # add 5 bonus points to every student, in one line
scaled = grades * 1.1   # a 10% scaling curve

print("Original grades:")
print(grades)
print()
print("Curved (+5):")
print(curved)
print()
print("Scaled (x1.1):")
print(scaled.round(1))
````

---
## 7. Boolean Filtering

Comparing a Series to a value produces a Series of `True`/`False` -- and you can use that directly to filter, keeping only the rows where the condition holds. This is the pattern you'll use constantly once real datasets show up in Lesson 04.

````python
grades = pd.Series([88, 92, 79, 95, 61], index=students)

passed_with_distinction = grades >= 90
print("Boolean mask (grades >= 90):")
print(passed_with_distinction)
print()

print("Filtered Series (only students scoring 90+):")
print(grades[passed_with_distinction])
print()

# The mask and the filter are usually written together, in one line
print("Same thing, written directly:")
print(grades[grades >= 90])
````

---
## 8. A First Look at Missing Values

Real-world data has gaps. Pandas represents a missing value as `NaN` ("Not a Number"), and `.isna()` / `.notna()` let you detect it. Module 06 covers handling missing data in depth -- this is just enough to recognize it when you see it.

````python
grades_with_gaps = pd.Series([88, np.nan, 79, np.nan, 61], index=students)

print("Series with missing values:")
print(grades_with_gaps)
print()
print("Which entries are missing?")
print(grades_with_gaps.isna())
print()
print(f"Count of missing values: {grades_with_gaps.isna().sum()}")
````

## ⚠️ Common Mistakes

````python
# -- Mistake 1: Using [ ] with an integer and expecting POSITION, not label -----

custom_index_series = pd.Series([10, 20, 30], index=[5, 6, 7])

print("Mistake 1 — plain bracket indexing follows the INDEX LABELS, not position:")
print(f"  custom_index_series[5]     = {custom_index_series[5]}   <- this is LABEL 5 (the first value, 10)")
print(f"  custom_index_series.iloc[0] = {custom_index_series.iloc[0]}   <- this is POSITION 0 (also 10, but for a clearer reason)")
print()
print("  When the index itself contains integers, plain [] indexing gets ambiguous.")
print("  Always prefer the explicit .loc (label) or .iloc (position) instead.")
````

````python
# -- Mistake 2: Assuming the index resets to 0..n after filtering ---------------

grades = pd.Series([88, 92, 79, 95, 61], index=students)
top_scorers = grades[grades >= 90]

print("Mistake 2 — filtering keeps the ORIGINAL labels, it doesn't renumber:")
print(top_scorers)
print()
try:
    top_scorers.iloc[2]
except IndexError as e:
    print(f"  Caught error: {e}")
print("  top_scorers only has 2 rows (Ben, Diya) -- .iloc[2] is out of range, even")
print("  though the ORIGINAL index went all the way up. Use .reset_index() if you")
print("  need fresh 0..n positions after filtering.")
````

````python
# -- Mistake 3: Forgetting arithmetic ALIGNS by index, not by position ----------

monday_sales = pd.Series([100, 200, 300], index=["apple", "banana", "cherry"])
tuesday_sales = pd.Series([10, 20, 30], index=["banana", "cherry", "date"])

total = monday_sales + tuesday_sales
print("Mistake 3 — adding two Series aligns by LABEL, producing NaN for unmatched labels:")
print(total)
print()
print("  'apple' and 'date' each appear in only ONE Series, so pandas has nothing")
print("  to add them to -- the result is NaN, not an error. Use .add(..., fill_value=0)")
print("  if you want missing labels treated as zero instead:")
print()
print(monday_sales.add(tuesday_sales, fill_value=0))
````

---
## 📌 Key Takeaways

- **A Series is a NumPy array with a label attached to every value** — that label (the index) is what lets pandas align, filter, and merge data by meaning instead of raw position.

- **`.loc` means "by label" and `.iloc` means "by position" — they are not interchangeable**, especially once you filter, sort, or otherwise end up with a non-sequential index.

- **Arithmetic between two Series aligns by index label, not position** — mismatched labels produce `NaN` rather than an error, and `.add(fill_value=0)` is the standard fix when you want missing labels treated as zero.
