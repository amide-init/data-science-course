## 1. Creating a DataFrame from a Dict of Lists

The most common way to build a DataFrame by hand: each **key** becomes a column name, and each **list** becomes that column's values, top to bottom.

````python
import pandas as pd
import numpy as np

student_data = {
    "name": ["Aisha", "Ben", "Chen", "Diya", "Elan"],
    "age": [20, 21, 19, 22, 20],
    "score": [88, 92, 79, 95, 61],
}

df = pd.DataFrame(student_data)
print(df)
print()
print(f"type: {type(df)}")
````

Notice the automatic 0, 1, 2... row index on the left, just like a Series -- because a DataFrame IS built from Series, and every Series needs an index.

---
## 2. Creating a DataFrame from a List of Dicts

Sometimes data naturally arrives one **row** at a time instead of one column at a time -- for example, one dict per API response or per form submission. Pandas handles this just as easily; each dict becomes a row, and the dict keys become column names.

````python
orders = [
    {"order_id": 1, "product": "Notebook", "quantity": 3, "price": 45},
    {"order_id": 2, "product": "Pen", "quantity": 10, "price": 10},
    {"order_id": 3, "product": "Eraser", "quantity": 5, "price": 5},
]

orders_df = pd.DataFrame(orders)
print(orders_df)
````

---
## 3. Creating a DataFrame from a NumPy Array

Since a DataFrame is built on NumPy underneath, you can also wrap a plain 2D array directly -- you just need to supply column names yourself, since a raw array has none.

````python
scores_array = np.array([
    [88, 92, 79],
    [95, 61, 70],
])

scores_df = pd.DataFrame(scores_array, columns=["Quiz1", "Quiz2", "Quiz3"])
print(scores_df)
````

---
## 4. Inspecting a DataFrame

Before doing anything else with a new dataset, these are the checks you'll run every single time -- they answer "how big is this, what's in it, and does it look right?"

````python
df = pd.DataFrame(student_data)

print(f"df.shape   : {df.shape}   <- (rows, columns)")
print(f"df.columns : {list(df.columns)}")
print(f"df.index   : {list(df.index)}")
print()
print("df.dtypes:")
print(df.dtypes)
````

````python
print("df.head(3)  -- first 3 rows:")
print(df.head(3))
print()
print("df.tail(2)  -- last 2 rows:")
print(df.tail(2))
````

````python
print("df.info() -- structure summary (columns, non-null counts, dtypes, memory):")
df.info()
````

````python
print("df.describe() -- summary statistics for NUMERIC columns only:")
print(df.describe())
````

---
## 5. Selecting Columns: Single Bracket vs Double Bracket

This is the detail that trips up almost everyone at first: **one** pair of brackets with a **single** column name returns a Series (1D); **double** brackets (a list of names, even a list of one) return a DataFrame (2D, still a table).

````python
single_column = df["score"]
print("df['score']  -- single brackets, ONE name -> a Series:")
print(single_column)
print(f"type: {type(single_column)}")
print()

single_column_as_df = df[["score"]]
print("df[['score']]  -- double brackets, a LIST of one name -> still a DataFrame:")
print(single_column_as_df)
print(f"type: {type(single_column_as_df)}")
print()

multiple_columns = df[["name", "score"]]
print("df[['name', 'score']]  -- multiple columns -> a DataFrame:")
print(multiple_columns)
````

---
## 6. Selecting Rows: `.loc` and `.iloc`

Just like a Series, `.loc` selects by label and `.iloc` selects by position -- except now every row is a whole record, not a single value.

````python
print("df.loc[2]  -- row at LABEL 2 (a Series: one row, columns become the index):")
print(df.loc[2])
print()

print("df.iloc[0:2]  -- rows at POSITIONS 0 and 1 (a DataFrame, end excluded):")
print(df.iloc[0:2])
print()

print("df.loc[1, 'score']  -- a single cell, by row label AND column name:")
print(df.loc[1, "score"])
````

---
## 7. Adding a New Column

New columns are usually **computed** from existing ones -- assign to a new column name, and pandas applies the calculation to every row at once (vectorized, just like Lesson 01's Series arithmetic).

````python
orders_df = pd.DataFrame(orders)

orders_df["total"] = orders_df["quantity"] * orders_df["price"]
print("After adding a computed 'total' column:")
print(orders_df)
````

---
## 8. Dropping a Column

`.drop()` removes a column (or row) -- but by default it targets **rows**, so dropping a column requires `axis=1` (or the equivalent `columns=` keyword).

````python
orders_no_price = orders_df.drop(columns=["price"])
print("After dropping the 'price' column:")
print(orders_no_price)
print()
print("Original orders_df is UNCHANGED (drop returns a new DataFrame by default):")
print(orders_df)
````

## ⚠️ Common Mistakes

````python
# -- Mistake 1: Confusing single-bracket (Series) with double-bracket (DataFrame) --

df = pd.DataFrame(student_data)

as_series = df["name"]
as_dataframe = df[["name"]]

print("Mistake 1 — these LOOK similar printed, but their types differ:")
print(f"  df['name']   -> {type(as_series).__name__}")
print(f"  df[['name']] -> {type(as_dataframe).__name__}")
print()
print("  This matters because a Series and a single-column DataFrame support")
print("  different operations -- e.g. .describe() on each gives very different output.")
````

````python
# -- Mistake 2: Forgetting axis=1 (or columns=) when dropping a column ----------

print("Mistake 2 — .drop() targets ROWS by default:")
try:
    df.drop("score")   # no axis specified -- pandas looks for a ROW labeled "score"
except KeyError as e:
    print(f"  Caught error: {e}")
print()
print("  Correct: df.drop('score', axis=1)  OR  df.drop(columns=['score'])")
print(df.drop(columns=["score"]))
````

````python
# -- Mistake 3: Chained indexing when assigning a value -------------------------

df = pd.DataFrame(student_data)

print("Mistake 3 — assigning through TWO separate bracket lookups is unreliable:")
print("  df[df['score'] < 70]['score'] = 70   # DON'T do this -- may silently fail")
print("  to modify the original df (pandas can't guarantee it's not a copy).")
print()
print("  Correct: use .loc with BOTH the row condition and column in ONE lookup:")
df.loc[df["score"] < 70, "score"] = 70
print(df)
````

## 📌 Key Takeaways

- **A DataFrame is a collection of Series sharing one row index** — it's the two-dimensional table structure you'll use for nearly every dataset in this course, and everything you learned about `.loc`, `.iloc`, and boolean filtering on a Series applies to its rows and columns too.

- **Single brackets (`df['col']`) return a Series; double brackets (`df[['col']]`) return a DataFrame** — the shape of your selection controls the type you get back, which matters for which methods are available next.

- **`.head()`, `.info()`, `.describe()`, and `.shape` are the first four things to run on any new dataset** — they tell you its size, structure, data types, and a statistical summary before you write a single line of analysis.
